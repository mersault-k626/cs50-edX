## 2.2 Reading Levels



### 2.2.1 The Problem

A reader's **reading level** corresponds, broadly, to the age or grade at which they are comfortable reading. The level is a function of how complex the text is — specifically, the complexity of the **words** and **sentences** involved. Simple, short phrasing maps to a young level; longer, more grammatically involved text maps to a higher one.

This complexity can be quantified. A **scientific measure** — a mathematical function that analyzes a body of text — can be applied to a passage to estimate the grade level at which it reads.



### 2.2.2 Worked Illustration

Three sample texts illustrate the range of levels the measure can detect:

| Sample text | Characteristics | Reading level |
|---|---|---|
| *"1 fish, 2 fish, red fish, blue fish."* | Very short phrases, little complexity | Kindergarten (before grade 1) |
| *"Congratulations. Today is your day. You're off to great places…"* | Rhyming, slightly more substance | Grade 3 |
| *"It was a bright cold day in April, and the clocks were striking 13…"* | More complex sentences and words, more grammar | Grade 10 |

As the words and sentence structure grow more complex, the computed grade level rises accordingly.



### 2.2.3 Why Strings Are Not Enough on Their Own

To apply any such metric, the text must first be **represented in memory** — and the natural representation is the **string**.

In Week 1, strings could effectively only be printed or displayed *wholesale*:

```c
string s = "1 fish, 2 fish, red fish, blue fish.";
printf("%s\n", s);
```

To *analyze* a body of text rather than merely display it, strings must be broken down at a **finer-grained level** — examining individual characters, words, and sentences. Reaching that granularity requires understanding how strings actually work internally: taking the "hood off the car" to see how the computer represents and manipulates them.

## 2.3 Cryptography (Preview)

**Cryptography** is the art of scrambling information — more precisely, **encrypting** it — so that secure communications can be sent. It underlies the everyday assumption that a **text message**, an **email**, or an **online credit-card checkout** is secure.



### 2.3.1 The Core Requirement: Reversibility

The central challenge is sending a message such that no one other than the intended recipient can read it — even if they **physically intercept** it in transit, which is all too possible in the digital world. This demands a way of scrambling information that is **reversible**:

1. **Encrypt** — the sender writes a message and scrambles it.
2. **Send** — the scrambled message travels, possibly passing through many other hands.
3. **Decrypt** — the recipient descrambles it back into the original.

An encrypted message therefore carries no usable meaning to an interceptor, yet can be perfectly recovered by the intended recipient. (For example, a scrambled string can be made to decrypt back to plaintext such as `"This is CS50"`.) The recurring open question for the coming weeks is *why* such schemes are — or are not — actually secure, and what additional tools are needed to implement them. (The mechanics are revisited in detail at the end of the lecture.)

## 2.4 Debugging: Motivation and Origins



### 2.4.1 Motivation

**Bugs** — mistakes in programs — are entirely commonplace. It is normal for an early C program to fail to compile or run at all until the code is corrected. Assistance tools exist (such as the virtual rubber duck at CS50.AI, also embedded into VS Code at cs50.dev), but a primary goal is to develop **lifelong, self-sufficient debugging skills**: the ability to find and fix problems without a duck, a TA, or any other person nearby. A range of techniques serves this single end — **finding and removing bugs** from software.



### 2.4.2 Origin of the Term "Bug"

The term *bug* was popularized by **Dr. Grace Hopper**, a **rear admiral** in the U.S. Navy and one of the original programmers of the **Harvard Mark I**, a very early mainframe computer (part of which is still on display in the lobby of Harvard's science and engineering complex).

The Mark I was succeeded by the **Harvard Mark II**. While working on the Mark II, Hopper and her team traced a problem with the machine to a **literal bug — a moth — caught inside the computer's circuitry**. They taped it into their logbook with the note:

> *"first actual case of bug being found."*

Ever since, **bug** and **debugging** have been the standard terms for, respectively, a defect in code and the process of finding and eliminating it.

## 2.5 Debugging in Practice

The tools below help locate and resolve mistakes without relying on outside help. They fall into two categories: errors the compiler catches, and errors only the running program reveals.



### 2.5.1 Syntax Errors vs. Logical Errors

A **syntax error** is a programming mistake that stops the code from compiling at all. It is a deal breaker: the compiler halts and refuses to produce an executable until the mistake is fixed. A **logical error** compiles cleanly but produces incorrect behavior at **run time** — the grammar of the code is valid, but it does not do what was intended.

The compiler understands only the C language and its required syntax. It has no knowledge of what a program is *supposed* to accomplish. CS50's **check50** can assess the *correctness* of a program automatically; the compiler can only enforce the rules of C.



#### 2.5.1.1 Common Syntax Errors Demonstrated with `buggy.c`

The following program is intentionally written with several bugs:

```c
main(void)
{
    printf("hello, world")
}
```

Compiling it with `make buggy` surfaces the mistakes one at a time:

| Bug | Symptom / Compiler Message | Fix |
| --- | --- | --- |
| Missing `#include <stdio.h>` | `error: call to undeclared library function 'printf'` | Add the header file declaring `printf`. |
| `studio.h` instead of `stdio.h` | Header not found / still undeclared | It is **st**andard **i**nput **o**utput: `stdio.h` (an annual FAQ). |
| Missing semicolon | `error: expected ';' after expression` | End the statement with `;`. |
| Missing `\n` | *No compiler error* — the next prompt lands on the same line | Add `\n` (an aesthetic/logical detail, not a syntax error). |

The header file matters because it tells the compiler the program plans to use functions declared therein, such as `printf`. Compilation stops on the first error and never reaches the program, so each error must be fixed in sequence.



#### 2.5.1.2 More Errors: Variables and the CS50 Library

Extending the program to read a name introduces a new set of mistakes:

```c
#include <stdio.h>

int main(void)
{
    name = get_string("What's your name? ");
    printf("hello, world\n");
}
```

| Bug | Compiler Message | Fix |
| --- | --- | --- |
| Variable type not declared | `error: use of undeclared identifier 'name'` | Declare the type. Text is not an `int`; it is a `string`. |
| `string` not recognized | `error: use of undeclared identifier 'string'; did you mean 'stdin'?` | Add `#include <cs50.h>`. |
| Wrong placeholder / variable | Prints literally `hello, world` instead of the name | Use the `%s` format code and pass the variable. |

Declaring a variable's type is mandatory in C (and C++, Java, and others), unlike Python. The `string` error is a **red herring**: the compiler suggests `stdin` simply because it is the closest-looking word it already knows. `string` is not a native C type — it is one of CS50's **training wheels**, declared in `cs50.h` along with `get_string`, `get_int`, and others. These training wheels let the early weeks ignore what a string really is; that layer is peeled back later, after which strings still exist but go by a different name.

The corrected program:

```c
#include <cs50.h>
#include <stdio.h>

int main(void)
{
    string name = get_string("What's your name? ");
    printf("hello, %s\n", name);
}
```

The earlier "wrong output" version was *not* a syntax error — the code was grammatically valid — but it was logically wrong, the kind of bug no error message reveals.



### 2.5.2 printf Debugging

When a logical bug compiles cleanly, there is no error message to read. **`printf`** becomes a diagnostic tool: beyond printing program output, it can *temporarily* print the values of variables to reveal what is happening at run time. Once the bug is understood, the temporary lines are deleted.

A loop intended to print a stack of 3 bricks instead prints 4:

```c
#include <stdio.h>

int main(void)
{
    for (int i = 0; i <= 3; i++)
    {
        printf("#\n");
    }
}
```

To diagnose, print `i` on each pass:

```c
for (int i = 0; i <= 3; i++)
{
    printf("# %i\n", i);   // temporary diagnostic
}
```

The output reveals the loop runs at `i = 0, 1, 2, 3` — four iterations. The fix is to stop *before* 3:

```c
for (int i = 0; i < 3; i++)   // exactly 3 bricks
```

Equivalently, counting could start at 1 and run `i <= 3`. The **canonical convention** is to start at 0 and go *up to but not through* the count — i.e., `i < n`.



#### 2.5.2.1 Abstracting into a Helper Function

The brick-printing logic can be abstracted into a helper function. Because it is defined *below* `main`, a **function prototype** (its first line plus a semicolon) must appear above `main` so the compiler knows it before use:

```c
#include <cs50.h>
#include <stdio.h>

void print_column(int height);   // prototype

int main(void)
{
    int h = get_int("Height: ");
    print_column(h);
}

void print_column(int height)
{
    for (int i = 0; i <= height; i++)   // same off-by-one bug, deliberately
    {
        printf("#\n");
    }
}
```

Abstraction makes the program modular but creates *more code to debug*, and the **edit–recompile–rerun cycle** of adding `printf` statements grows tedious. This motivates a dedicated debugger.



### 2.5.3 debug50 — Step-Through Debugging

**`debug50`** is a CS50 command that launches an industry-standard debugger — software that lets execution be slowed or paused and walked through line by line. The wrapper exists only because starting the debugger in VS Code is otherwise tedious; everything after launch is standard practice, not CS50-specific.



#### 2.5.3.1 Breakpoints

Running the debugger without a breakpoint produces:

```text
Looks like you haven't set any breakpoints. Set at least one breakpoint
by clicking to the left of a line number and then re-run debug50.
```

A **breakpoint** tells the debugger where to pause execution. Hovering to the left of a line number shows a small red dot; clicking fixes the breakpoint. A breakpoint at `main` — the program's entry point — pauses execution right away.

```bash
debug50 buggy
```



#### 2.5.3.2 What the Debugger Shows

When execution pauses, the line is highlighted in yellow but has **not yet executed** — which is why no output appears yet. Three areas are notable:

- **Variables pane** (left): the live value of any variable in memory.
- **Call stack** (bottom): which function execution is paused inside (e.g., `main` in `buggy.c`).
- **Control icons** (top): described below.



#### 2.5.3.3 Garbage Values

Before a variable is assigned, the Variables pane may show a meaningless value — for example, `h` showing **32764**. This is a **garbage value**: a leftover default that exists because that region of memory (**RAM**) was previously used for something else. The assigning line overwrites it. This has real security implications — leftover memory could just as easily contain a password — and a debugger (let alone C itself) makes it unsettlingly easy to inspect memory. A value is "garbage" in the sense that the programmer did not put it there.



#### 2.5.3.4 The Control Icons

| Icon | Name | Behavior |
| --- | --- | --- |
| Play button | **Continue** | Resume normal execution; stop stepping. |
| Curved arrow over a dot | **Step over** | Execute the current line, then pause on the next — *without* descending into any function it calls. |
| Arrow pointing in | **Step into** | Descend *into* a called function and continue line by line inside it. |



#### 2.5.3.5 Stepping Through the Bug

Stepping confirms the off-by-one bug directly: stepping over `get_int` and typing `3` sets `h` to 3; **stepping into** `print_column` pauses on its `for` line (where `i` first shows garbage); stepping over makes `i` become 0, then prints `#` on each pass for `i = 0, 1, 2, 3`. At `i = 3`, the fourth hash appears because `3 <= 3` is true. The debugger does not solve the bug — human reasoning is still required — but it lets execution be paused and examined at one's own pace.



#### 2.5.3.6 Order of `for`-Loop Execution

The exact sequence, worth stating since the debugger highlights whole lines:

1. **Initialization** runs **once** (clause before the first semicolon).
2. The **condition** (Boolean expression) is checked.
3. If true, the **body** executes.
4. The **update** runs (e.g., `i++`).
5. Return to step 2 — until the condition is false.

The three clauses must appear in the fixed order **initialization; condition; update** and cannot be reordered, though each may contain multiple statements. The start-at-0, `< n` form is canonical.

## 2.6 Debugging Case Studies with debug50

Two worked examples show how `debug50` localizes bugs in practice — one using **step over**, the other requiring **step into**.



### 2.6.1 Case Study: `guess.c` — Step Over Reveals a Loop Bug

A guessing game (guess a number, get told right or wrong) misbehaves: running `./guess` and entering a guess produces an endless flood of `Wrong guess!` messages.

```c
#include <cs50.h>
#include <stdio.h>

int main(void)
{
    int number = 5;                       // line 6
    int guess = get_int("Guess: ");       // line 8

    while (guess != number)               // line 10  ← bug
    {
        printf("Wrong guess!\n");
    }
}
```

**Diagnosing with `debug50 guess`:**

1. A **breakpoint** is set (click left of a line number → solid red dot).
2. At the pause, the Variables pane shows `number` as `0` and `guess` as a **garbage value** (`32764`) — neither line has executed yet.
3. **Step over** line 6 → `number` becomes `5`, as expected.
4. **Step over** line 8 → prompted for a guess; entering `3` makes `guess` become `3`.
5. **Step over** through the loop body → `Wrong guess!` prints, which seems right — but the highlighted line then jumps **back up to line 10** and prints `Wrong guess!` again, and again.

**The bug:** a `while` loop was used where the condition should be checked only **once**. Since nothing inside the loop ever changes `guess`, the condition `guess != number` stays true forever — an infinite loop. The fix is an **`if`**:

```c
if (guess != number)
{
    printf("Wrong guess!\n");
}
```

**A second, lingering bug:** after that fix, guessing `3` now prints **both** `Wrong guess!` *and* `You're correct!`. The "correct" branch is not properly guarded against the "wrong" case (the two outcomes are not mutually exclusive). Tracing this one with `debug50` is left as an exercise — it is the kind of logical bug the debugger is built to expose.



### 2.6.2 Case Study: `secret.c` — Step Into Reveals a String-Comparison Bug

A password program should print `Come on in` when the correct phrase (`please`) is entered — but it never does, even with the right phrase.

```c
#include <cs50.h>
#include <stdio.h>
#include <string.h>

bool check_phrase(string phrase);

int main(void)
{
    string phrase = get_string("Secret phrase: ");
    bool correct = check_phrase(phrase);
    if (correct)
    {
        printf("Come on in\n");           // line 13
    }
}

bool check_phrase(string phrase)
{
    string password = "please";           // line 22
    if (phrase == password)                // bug: comparing strings with ==
    {
        return true;
    }
    return false;                          // line 29
}
```

**Diagnosing with `debug50 secret`:**

1. Stepping **over** the prompt and entering `please` sets `phrase` to `"please"`; `correct` is `false`.
2. Stepping **over** `check_phrase` leaves `correct` *still* `false` — so line 13 is skipped and nothing prints. Step-over alone does not reveal *why*.
3. Re-running and using **step into** on `check_phrase` descends inside the function. `password` is set to `"please"` (line 22). Stepping over the comparison `if (phrase == password)`, execution jumps to line 29 (`return false`) — the condition was false **even though both strings are `"please"`**.

**The bug:** in C, two strings **cannot** be compared with `==`. Doing so does not compare their characters. The fix is the **`strcmp`** function (string compare) from `string.h`:

```c
if (strcmp(phrase, password) == 0)
{
    return true;
}
```

`strcmp(a, b)` returns **`0`** when the two strings are exactly equal. After this change, entering `please` correctly prints `Come on in`.

> **Key takeaway:** comparing strings in C requires `strcmp(a, b) == 0`, never `a == b`. And `debug50` is not only for stepping *over* code — **stepping into** a function exposes bugs hidden beneath the surface of a call.

## 2.8 What Compiling Really Means



### 2.8.1 The Mental Model

When **source code** is written in a language like C, the computer cannot understand it natively — computers understand only **zeros and ones**, also known as **machine code**. A **compiler** is the program that converts source code into machine code. More generally, a compiler translates **one language into another**; here, from C to machine code.



### 2.8.2 `make` Is Not the Compiler

The two commands typically run after creating a file such as `hello.c` are:

```bash
make hello
./hello
```

It was a "white lie" to imply that the compiler is called `make`. **`make` is a command that automates compilation** — it makes the program by compiling it, but it is not itself the compiler. The actual compiler is **`clang`** (for *C language*), a popular, freely available, open-source compiler.



### 2.8.3 Running `clang` Directly, `a.out`, and Command-Line Arguments

`make` essentially automates a direct `clang` invocation:

```bash
clang hello.c
```

The **default output filename** from `clang` is not `hello` but **`a.out`** — standing for **assembler output**, a name retained for historical reasons. The program then runs with `./a.out`.

**Command-line arguments** are the words typed after a program's name:

| Command | Program | Command-line argument(s) |
|---|---|---|
| `code hello.c` | `code` | `hello.c` |
| `make hello` | `make` | `hello` |
| `clang hello.c -lcs50` | `clang` | `hello.c`, `-lcs50` |

When a program uses the CS50 library, a bare `clang hello.c` fails:

```text
Linker command failed with exit code 1 ... undefined reference to get_string
```

This is distinct from an *undeclared* error caused by a missing `#include` — that line is present. The compiler simply does not know **where to find** CS50's implementation of `get_string`. The fix is to **link in** the library with `-lcs50`, and the output can be named with `-o`:

```bash
clang -o hello hello.c -lcs50
```

Running such commands by hand is tedious — the flags and order must all be remembered. This is exactly what **`make` automates behind the scenes**: the `-o` output name, the source file, and `-lcs50`.



### 2.8.4 The Four Steps of Compiling

"Compiling" is the industry term of art for what is technically **four separate processes**, all happening automatically in succession:

| Step | Name | What it does |
|---|---|---|
| 1 | **Preprocessing** | Resolves `#` directives; copies header-file contents into the code |
| 2 | **Compiling** | Translates preprocessed C into **assembly code** |
| 3 | **Assembling** | Converts assembly into **zeros and ones** (machine code) |
| 4 | **Linking** | Combines your machine code with the libraries' machine code into one executable |

**Step 1 — Preprocessing.** Lines beginning with `#` (such as `#include`) are **preprocessor directives**. The preprocessor finds the named **header file** on disk (e.g., `cs50.h`, `stdio.h`) and effectively **copies and pastes its contents** into the code. What the headers supply are **prototypes** — declarations of a function's inputs and return value. For example, `get_string`'s prototype (in `cs50.h`) takes a *prompt* as input and returns a *string*; `printf`'s prototype lives in `stdio.h`. (Analogy: `#include` is like *copy/paste*, whereas `#define` is like *find/replace*.)

**Step 2 — Compiling.** The preprocessed C is translated into **assembly code** — the lowest-level instructions the **CPU** understands (made by Intel, AMD, Apple, etc.). Assembly is cryptic but contains recognizable names like `main`, `printf`, and `get_string` alongside instructions such as `mov`, `push`, `call`. It is how programmers wrote code before C, Python, and Java existed (and before that, raw zeros and ones via **punch cards**). Because assembly is the language of a specific CPU, it is **CPU-specific** — which is also why `a.out` is named "assembler output."

**Step 3 — Assembling.** Assembly is converted into the actual **zeros and ones** (machine code). The code in `hello.c` becomes machine code — but your code is not the entire program.

**Step 4 — Linking.** Your machine code is combined with the already-compiled machine code of the libraries used — the **CS50 library** and the **standard I/O** implementation (which supplies `printf`). **Linking** combines all these chunks of zeros and ones into one larger executable. This is what `-lcs50` enables; its absence caused the earlier *linking* error.



### 2.8.5 Why "Compiling" Covers All Four — and Why the Complexity Exists

Programmers treat all four steps collectively as just **"compiling."** The layered design exists for good reasons:

- **Independent implementation and collaboration**: different people can build the preprocessor, compiler, assembler, and linker separately.
- **Multiple back-ends from one source**: different compilers can output assembly for different targets — iPhone, Android, Mac, PC — while everyone writes in the *same* high-level language. The same C source yields different assembly per CPU, which is why a Mac binary will not run on a PC and vice versa.

## 2.9 Decompiling and Reverse Engineering

Since compiling goes from source code to machine code, the natural question is whether the process can be **reversed** — taking someone's machine code (e.g., Microsoft Word or Excel) and reconstructing the source. Because the zeros and ones must be **accessible** to the user's computer in order to run, this is a **genuine intellectual-property threat** and arises in matters of law and IP.



### 2.9.1 Why It Is Hard

- **Architecture knowledge and painstaking decoding**: the target **CPU/architecture** must be known, and the zeros and ones must be decoded by hand in chunks of **8, 16, 32, or 64 bits**. A glance at a pattern of bits gives no hint that it prints `hello, world`. The effort is so large that it would almost certainly **take less time to simply rewrite the software the normal "forward" way** than to reverse engineer it.

- **Irrecoverable information**: some details are permanently lost in compilation. For example, a **`for` loop** and a **`while` loop** are functionally equivalent and a purely **stylistic choice**; both compile to the *same* pattern of zeros and ones. It is therefore impossible to tell from the machine code which one the original source used. The original source cannot be perfectly reconstructed.



### 2.9.2 Teaser: Not All Languages Are Compiled

Not all languages work this way. **JavaScript** source code is **sent from web servers to web browsers**, and the source of any website (e.g., harvard.edu, gmail.com) can be viewed directly in the browser. In such cases the source code is **executed directly** by the underlying computer rather than compiled into machine code.

## 2.10 Memory and Data Types

Every value stored in a variable occupies physical space in memory. Each C data type reserves a fixed number of bytes (with the exception of `string`, whose size depends on its contents).



### 2.10.1 Data-Type Sizes

| Type | Size | Notes |
|------|------|-------|
| `bool` | 1 byte (8 bits) | Only needs 1 bit (true/false), but a whole byte is used because it is simpler and more efficient; 7 bits are effectively unused. |
| `int` | 4 bytes (32 bits) | ~4 billion possible values. If signed, roughly −2 billion to +2 billion (the range is split to allow negatives). |
| `long` | 8 bytes (64 bits) | Twice the size of an `int`; vastly more values (on the order of ~9 quintillion signed) — far beyond an `int`'s range. |
| `float` | 4 bytes | Real numbers with decimal points; limited precision (a finite number of significant digits). |
| `double` | 8 bytes | Twice the bits of a `float`, so greater precision. |
| `char` | 1 byte | A single ASCII character. |
| `string` | varies | Depends on length, including the terminating byte. `"HI!"` is ~4 bytes; `"hello"` is ~6 bytes. |

> **Floating-point imprecision** persists regardless of whether a `float` or a `double` is used. A `double` only narrows the error; it does not eliminate it. This remains a fundamental problem for scientific, financial, and other precision-critical computing.



### 2.10.2 Physical Memory

Memory physically lives on a stick of RAM called a **DIMM**, which carries small black chips full of electronic zeros and ones. Conceptually, memory can be modeled as a **grid of bytes**, where each byte has a numeric **address** — byte 0, byte 1, byte 2, and so on, up to (for example) ~1 billion for 1 GB.

Key properties of how values sit in memory:

- A value occupies **contiguous** bytes — back to back. A 4-byte `int` uses four adjacent bytes; the bits cannot be scattered across arbitrary locations.
- The notions of "top," "bottom," "left," and "right" are irrelevant. The hardware is simply a long sequence of addressable bytes.
- Storing a value means **permuting** physical bits — turning specific zeros and ones on or off to represent that value. A 32-bit `int` is just a particular pattern of 32 bits.

## 2.11 From Variables to Arrays



### 2.11.1 Three Separate Variables

A naive program to average three problem-set scores (72, 73, 33) declares one variable per score:

```c
#include <stdio.h>

int main(void)
{
    int score1 = 72;
    int score2 = 73;
    int score3 = 33;

    printf("Average: %i\n", (score1 + score2 + score3) / 3);
}
```

Each `int` occupies 4 bytes, so the three variables use 12 contiguous bytes total. The compiler decides where to place them.



### 2.11.2 The Integer-Division (Truncation) Bug

The program prints `59`, but the true average is `59.33…`. Because all operands are integers, **integer division truncates** — everything after the decimal point is discarded. (This is the same truncation issue seen at the end of Week 1.)

Fixing it requires two coordinated changes:

1. **Print as a float.** Change the format code from `%i` to `%f`. (`%f` works for both `float` and `double`.) This alone is *not* sufficient — passing an integer to `%f` produces a compiler error: *"format specifies double, but the argument has type int."*
2. **Make the arithmetic floating-point.** Several options:
   - Convert the operands to floats, e.g. write `72.0`, `73.0`, `33.0`.
   - Convert just the denominator: `… / 3.0`. Involving **even one** floating-point value **promotes** the entire expression to floating point — not all operands need converting.
   - **Cast** the divisor: `… / (float) 3`. A cast in parentheses converts one data type to another.

```c
printf("Average: %f\n", (score1 + score2 + score3) / 3.0);
```

> **Notes on literals and format codes.** A literal like `3.0` defaults to a **double** (64 bits); raw decimal-point numbers are treated as doubles by the compiler. Both `%i` and `%d` print integers (`%d` is "decimal integer"), but CS50 prefers **`%i`**. Format codes are inconsistent by historical accident — e.g., `%lli` prints a `long`. Naming a single-use intermediate variable `sum` (rather than something opaque like `z`) aids readability, even if creating a variable used only once is debatable design.



### 2.11.3 Why This Is Bad Design

Correct does not mean well designed. Declaring a fresh variable for each new score does not scale — by the end of a semester there could be 10, 20, or more scores, each requiring a new variable typed manually.

> **Heuristic — think about the extreme.** If there were not 3 scores but 30 or 300, would this approach still make sense? When the answer is clearly "no," a better-designed solution almost certainly exists in a well-designed language.



### 2.11.4 Arrays

An **array** is a chunk of **contiguous** memory holding many values of a single type, back to back. To store 3 values, request a chunk of size 3; for 30, request size 30; and so on.

**Declaration syntax** specifies the type, a name, and (in square brackets) the count:

```c
int scores[3];
```

This single line reserves 12 bytes (three 4-byte ints), all referable by the name `scores`.

**Indexing and assignment** use square brackets, and indices start at **0**:

```c
int scores[3];
scores[0] = 72;
scores[1] = 73;
scores[2] = 33;

printf("Average: %f\n", (scores[0] + scores[1] + scores[2]) / 3.0);
```

In memory the values remain contiguous. On screen they may appear to wrap to the next row, but that wrapping is purely a visual artifact — the machine has no up/down/left/right, only a sequence of addressable bytes. To store a different number of values, change only the number inside the brackets.



### 2.11.5 Making It Dynamic with a Loop

Reading scores interactively with `get_int` (from the CS50 header) avoids hard-coding them. Repeating three `get_int` calls is repetitive, so a `for` loop collapses them:

```c
#include <cs50.h>
#include <stdio.h>

int main(void)
{
    int scores[3];
    for (int i = 0; i < 3; i++)
    {
        scores[i] = get_int("Score: ");
    }

    printf("Average: %f\n", (scores[0] + scores[1] + scores[2]) / 3.0);
}
```

As `i` increments from 0 to 2, each input lands in the next array slot.



### 2.11.6 The Magic-Number Problem

The literal `3` is hard-coded in **two** places — the array size and the loop bound — yet both represent the same concept. Keeping them in sync is left to the honor system. This is a **magic number**.

A first fix uses one variable in both places, but better still, a value that should never change is declared **`const`**, and by convention constants are **capitalized**:

```c
const int N = 3;
int scores[N];
for (int i = 0; i < N; i++)
{
    scores[i] = get_int("Score: ");
}
```



### 2.11.7 Passing Arrays to Functions

Averaging logic can be factored into its own function. The crucial limitation:

> **C arrays do not know their own length.** Unlike Python or Java — where an array/vector can report its own length — a C function receiving an array cannot determine its size. The length must be passed as a **separate argument**.

When **defining** a function that accepts an array, specify the element type and a name, then use **empty square brackets** — do *not* put a size inside them. If `average` is defined **below** `main`, a **prototype** (its first line plus a semicolon) must appear above `main`:

```c
#include <cs50.h>
#include <stdio.h>

float average(int length, int array[]);   // prototype

int main(void)
{
    const int N = 3;
    int scores[N];
    for (int i = 0; i < N; i++)
    {
        scores[i] = get_int("Score: ");
    }
    printf("Average: %f\n", average(N, scores));   // pass array by name + its length
}

float average(int length, int array[])
{
    int sum = 0;
    for (int i = 0; i < length; i++)
    {
        sum += array[i];               // tightened form of sum = sum + array[i]
    }
    return sum / (float) length;       // cast to avoid integer truncation again
}
```

Points to note:

- The return type is `float` (a `double` would be more precise but is excessive here).
- The function is **called** by passing the array's **name** (no brackets) plus its length: `average(N, scores)`.
- The final `return` is again vulnerable to **integer division**: `sum / length` would truncate, so `(float)` casting (or making `sum` a float) forces floating-point arithmetic.

Running this with inputs 72, 73, 33 yields `59.333…` — enough precision for a grade, with only the usual floating-point imprecision remaining.

## 2.12 Functions

Until now, every program has been written entirely inside `main`. That is fine for short programs, but as programs grow to hundreds, thousands, or tens of thousands of lines, cramming everything into `main` becomes unwieldy and hard to navigate. C — and nearly every language developed since — allows code to be organized into **functions**.

> **Synonyms.** A function is also called a **procedure**, a **method** (especially in object-oriented languages), or a **subroutine**. All refer to the same idea.



### 2.12.1 What Is a Function?

A function is a **black box** with a set of **zero or more inputs** and **a single output**.

```
   a     b     c          (inputs / arguments)
   │     │     │
   ▼     ▼     ▼
┌─────────────────┐
│  func(a, b, c)  │        (the black box)
└─────────────────┘
         │
         ▼
         z                 (output / return value)
```

Concrete examples:

| Call | Inputs | Output |
|------|--------|--------|
| `add(3, 6, 7)` | 3, 6, 7 | 16 |
| `mult(4, 5)` | 4, 5 | 20 |



### 2.12.2 Why "Black Box"?

If a function is written by someone else, *how* it works internally does not matter — only that its output is what the name promises. `printf`, for instance, is used constantly but was never written by its users.

The same `mult(a, b)` could be implemented two completely different ways, with identical results:

```text
mult(a, b):              mult(a, b):
    output a * b             set counter to 0
                             repeat b times:
                                 add a to counter
                             output counter
```

`mult(3, 5)` returns 15 either way — the implementation differs, the output does not. This is the **contract** of using functions: behavior is predictable from the name. That is precisely why functions should have **clear, meaningful names** and be **well-documented**.



### 2.12.3 Why Use Functions?

| Benefit | Explanation |
|---------|-------------|
| **Organization** | Break a complicated problem into manageable sub-parts. |
| **Simplification** | A 10-line function is far easier to write and debug than a 1,000-line `main`. |
| **Reusability** | Write once, use as often as needed. `printf` was written once, decades ago, and has been reused ever since. |



### 2.12.4 Three Steps to Creating a Function

Creating and using a function involves **declaration**, **definition**, and the **call**.



#### 2.12.4.1 Declaration

A **declaration** gives the compiler a heads-up that a user-written function will appear later, so it does not choke on an unfamiliar name. Declarations go **atop the code, before `main`**. (This is exactly the **prototype** seen earlier with `print_column` and `average` — "declaration" and "prototype" are the same thing.)

Every declaration follows one standard form with three parts:

```c
return-type name(argument-list);
```

- **return-type** — the type of variable the function outputs.
- **name** — what the function is called (least important to the compiler, most important to humans — avoid `f`, `g`, `mystery`).
- **argument-list** — the comma-separated inputs, each with a **type** and a **name**.

A function to add two integers (the sum of two ints is an int):

```c
int add_two_ints(int a, int b);
```

A function to multiply two real numbers (the product of two reals is a real):

```c
float mult_two_reals(float x, float y);
// double mult_two_reals(double x, double y);  // equally valid, more precise
```

Argument names like `a`, `b`, `x`, `y` are fine when they carry no special meaning; use descriptive names when they do.



#### 2.12.4.2 Definition

The **definition** fills in the black box so that predictable behavior happens. Its first line is **almost identical** to the declaration — same return type, name, and argument list — but with **no semicolon**, followed by curly braces. The **`return`** statement passes the output back out.

`mult_two_reals`, two equivalent ways:

```c
float mult_two_reals(float x, float y)
{
    float product = x * y;
    return product;
}
```
```c
float mult_two_reals(float x, float y)
{
    return x * y;       // no need for an intermediate variable
}
```

`add_two_ints`, similarly:

```c
int add_two_ints(int a, int b)
{
    int sum = a + b;    // calculate the sum
    return sum;         // give the result back
}
```

A deliberately **bad** implementation still satisfies the contract — it adds correctly but is poorly designed (it loops, incrementing a counter, handling positive and negative `a` and `b` separately):

```c
int add_two_ints(int a, int b)
{
    int sum = 0;
    if (a > 0)
        for (int i = 0; i < a; sum++, i++);
    else
        for (int i = a; i < 0; sum--, i++);
    if (b > 0)
        for (int i = 0; i < b; sum++, i++);
    else
        for (int i = b; i < 0; sum--, i++);
    return sum;
}
```

It works, but illustrates the point: the caller does not care what happens inside the box, only that the output is correct. (This is a poorly designed box.)



#### 2.12.4.3 The Call

To **call** a function, pass it arguments of the expected types and assign its return value to something of the correct type. In `adder.c`:

```c
#include <cs50.h>
#include <stdio.h>

int add_two_ints(int a, int b);   // declaration / prototype

int main(void)
{
    int x = get_int("Give me an integer: ");
    int y = get_int("Give me another integer: ");
    int z = add_two_ints(x, y);   // the call: x and y are ints, z receives an int
    printf("The sum of %i and %i is %i\n", x, y, z);
}

int add_two_ints(int a, int b)
{
    int sum = a + b;
    return sum;
}
```

`x` and `y` are integers matching the two-int argument list, and the integer return value is stored in the integer `z`. Running it: inputs 10 and 20 yield 30; inputs −10 and 17 yield 7.



### 2.12.5 Miscellany: `void`

- A function that takes **no inputs** is declared with a **`void` argument list**. The most familiar example is `main(void)`.
- A function that produces **no output** is declared with a **`void` return type**. `printf` is the canonical example — it has a side effect (printing) but returns no usable value.



### 2.12.6 Practice Problem: `valid_triangle`

**Task:** write `valid_triangle`, taking three real numbers (the side lengths of a triangle) and returning **`true`** or **`false`** depending on whether those lengths can form a triangle.

Two geometric rules apply:
1. A triangle's sides must all have **positive** length.
2. The **sum of any two sides must be greater than the third** (e.g., 1, 2, 4 cannot form a triangle, since 1 + 2 is not greater than 4).

Since the output is true/false, the return type is `bool`:

```c
bool valid_triangle(float x, float y, float z);

bool valid_triangle(float x, float y, float z)
{
    // check for all positive sides
    if (x <= 0 || y <= 0 || z <= 0)
    {
        return false;
    }

    // check that the sum of any two sides exceeds the third
    if ((x + y <= z) || (x + z <= y) || (y + z <= x))
    {
        return false;
    }

    // if both tests passed, the sides form a valid triangle
    return true;
}
```

## 2.13 Variable Scope

Once functions enter the picture, **scope** becomes important. Scope is a characteristic of a variable that defines **from which functions that variable may be accessed**. (Week 1 introduced a narrower form of this — a variable declared inside a loop's curly braces exists only there. Function-level scope generalizes the idea.)

C has two primary scopes:

| Scope | Where declared | Who can access it |
|-------|----------------|-------------------|
| **Local** | Inside a function | Only the function in which it was created |
| **Global** | Outside *all* functions | Any function in the program |



### 2.13.1 Local Variables

Almost all variables used so far have been **local**. A local variable belongs to exactly one function:

```c
int triple(int x);   // declaration

int main(void)
{
    int result = triple(5);
}

int triple(int x)
{
    return x * 3;
}
```

Here `x` is **local to `triple`** — `main` cannot refer to it. `result` is **local to `main`** — `triple` cannot refer to it. Each name is visible only inside its own function.



### 2.13.2 Global Variables

A variable declared **outside of every function** is **global**, and any function may read or modify it:

```c
#include <stdio.h>

void triple(void);

float global = 0.5050;   // global — declared outside all functions

int main(void)
{
    triple();                  // modifies the global
    printf("%f\n", global);    // prints the *modified* value
}

void triple(void)
{
    global *= 3;
}
```

Both `main` and `triple` can see `global`. `main` calls `triple`, which sets `global` to `0.5050 × 3 ≈ 1.5150`; `main` then prints **1.5150**, not the original 0.5050.

> **Caution.** Globals are flexible — every function can share the value — but dangerous: one function may change the value before another expects it to change. They should be used sparingly.



### 2.13.3 Why the Distinction Matters: Pass by Value

In C, local variables are **passed by value** in function calls. When a variable is passed by value, the **callee** (the function receiving the input) gets its **own copy** of the variable, not the variable itself.

The consequence: a variable in the **caller** (the function making the call) is **unchanged unless it is overwritten**.

**No effect** — `triple(foo)` operates on a copy, so `foo` stays 4:

```c
int main(void)
{
    int foo = 4;
    triple(foo);        // triple works on a COPY; main's foo is untouched
}

int triple(int x)
{
    return x *= 3;
}
```

**Overwrites `foo`** — the return value is assigned back, so `foo` becomes 12:

```c
int main(void)
{
    int foo = 4;
    foo = triple(foo);  // triple's result (12) is stored back into foo
}

int triple(int x)
{
    return x *= 3;
}
```

The only way to reflect the change in a local caller variable is to **assign the function's return value back to it**.



### 2.13.4 Same Name, Different Scope

Reusing a variable name across functions is perfectly legal **as long as the variables live in different scopes** — and it happens often in collaborative, real-world code, especially for common names like `x`, `y`, `i`, `j`. It can, however, make tracing tricky. Consider:

```c
int increment(int x);

int main(void)
{
    int x = 1;
    int y;
    y = increment(x);
    printf("x is %i, y is %i\n", x, y);
}

int increment(int x)
{
    x++;
    return x;
}
```

There are **two distinct variables named `x`** — one local to `main`, one local to `increment`. When `main` calls `increment(x)`, `increment` does **not** receive `main`'s `x`; it receives a **copy**. It increments only its own copy and returns it.

Thinking of them as `x_m` (main's) and `x_i` (increment's) clarifies the trace:

1. `x_m = 1`
2. `y = increment(x_m)` — a copy of `x_m` (value 1) becomes `x_i`
3. inside `increment`: `x_i++` → `x_i` is 2; `return x_i` hands back 2
4. `y` receives 2; **`x_m` was never touched** and is still 1

Output:

```text
x is 1, y is 2
```

## 2.14 Arrays in Depth

Arrays were introduced earlier as a contiguous block of memory for many values of one type (see *From Variables to Arrays*). This section deepens that treatment with several properties that matter once arrays are passed around and manipulated.



### 2.14.1 The Post Office Analogy

An array maps neatly onto a bank of post office boxes:

| Array | Post office boxes |
|-------|-------------------|
| An **array** is a block of contiguous space in memory… | A **mail bank** is a large space on the post office wall… |
| …partitioned into small, identically-sized blocks called **elements**… | …partitioned into small, identically-sized blocks called **boxes**… |
| …each storing a certain amount of **data**… | …each holding a certain amount of **mail**… |
| …all of the same data type (`int`, `char`, …) | …all of a similar type (letters, small packages) |
| …accessed directly by an **index**. | …accessed directly by a **box number**. |



### 2.14.2 Indexing and Out-of-Bounds Access

In C, array elements are **indexed from 0**, not 1 — this is a major reason computer scientists count from zero. For an array of `n` elements, the first element is at index `0` and the last at index `n − 1`. A 50-element array spans indices `0` through `49`.

**C does not check array bounds.** It will not stop a program from accessing index `-3` or index `59` of a 50-element array. The code still *compiles*, but at run time, reading or writing memory outside the array may trigger the dreaded **segmentation fault** (a crash from touching memory the program was not given). This leniency demands care.



### 2.14.3 Declaration

An array declaration has three parts — **type**, **name**, and **size** (in square brackets):

```c
type name[size];

int    student_grades[40];   // 40 ints
double menu_prices[8];       // 8 doubles
```



### 2.14.4 An Element Behaves Like a Variable

A single element of an array of type `T` behaves exactly like any variable of type `T`, so all the familiar operations apply:

```c
bool truthtable[10];

truthtable[2] = false;          // assign
if (truthtable[7] == true)      // test in a condition
{
    printf("TRUE!\n");
}
truthtable[10] = true;          // DANGER: valid indices are 0–9; index 10 is out of bounds
```



### 2.14.5 Initialization Syntax

When declaring **and** initializing at once, a curly-brace list fills the array with starting values — far less tedious than assigning element by element:

```c
// instantiation syntax
bool truthtable[3] = { false, true, true };

// equivalent, element by element
bool truthtable[3];
truthtable[0] = false;
truthtable[1] = true;
truthtable[2] = true;
```

With the brace-initializer form, the **size may be omitted** — the compiler counts the elements:

```c
bool truthtable[] = { false, true, true };   // compiler infers size 3
```



### 2.14.6 Multi-Dimensional Arrays

An array may have more than one dimension — as many size specifiers as desired:

```c
bool battleship[10][10];   // a 10×10 grid
```

This is convenient for visualizing game boards (Battleship, Tic-Tac-Toe). **In memory, however, it remains a single one-dimensional array** — here, 100 elements back to back. Multiple dimensions simply multiply together; they are a human-facing **abstraction** for organization and perception, not a different memory layout.



### 2.14.7 Whole Arrays Cannot Be Copied with `=`

Individual elements can be treated as variables, but an **entire array cannot**. One array cannot be assigned to another with the assignment operator:

```c
int foo[5] = { 1, 2, 3, 4, 5 };
int bar[5];

bar = foo;        // ILLEGAL — not valid C
```

To copy the contents, each element must be copied **one at a time**, with a loop:

```c
int foo[5] = { 1, 2, 3, 4, 5 };
int bar[5];

for (int j = 0; j < 5; j++)
{
    bar[j] = foo[j];   // now bar is {1, 2, 3, 4, 5}
}
```

(Some more modern languages permit the simple `=` copy; C does not.)



### 2.14.8 Arrays Are Passed by Reference

The Variable Scope notes established that most variables in C are **passed by value** — the callee receives its own copy, so the caller's variable is unchanged. **Arrays are the exception: they are passed by reference.** The callee receives the **actual array**, not a copy — so changes made by the callee persist in the caller.

Intuitively, this makes sense: copying an array of thousands or millions of elements on every function call would be wasteful, so the array itself is handed over.

```c
void set_array(int array[4]);
void set_int(int x);

int main(void)
{
    int a = 10;
    int b[4] = { 0, 1, 2, 3 };
    set_int(a);       // passes a COPY of a
    set_array(b);     // passes the ACTUAL array b
    printf("%d %d\n", a, b[0]);
}

void set_array(int array[4])
{
    array[0] = 22;    // modifies main's b directly
}

void set_int(int x)
{
    x = 22;           // modifies only the local copy
}
```

Output:

```text
10 22
```

`a` stays **10** — `set_int` changed only its own copy (pass by value). But `b[0]` becomes **22** — `set_array` modified `main`'s actual array (pass by reference). *Why* arrays behave this way is revisited in a later week.

## 2.15 Strings Are Arrays of Characters



### 2.15.1 Chars: Single Quotes vs. Double Quotes

A single character is stored in a variable of type `char`. Character literals are written with **single quotes**, while string literals are written with **double quotes**.

```c
char c1 = 'H';
char c2 = 'I';
char c3 = '!';
```

When a program declares only these three `char` variables, they are typically stored **back to back** (contiguously) in memory. Printing them with the `%c` format code (`c` for character) outputs them as text:

```c
printf("%c%c%c\n", c1, c2, c3);   // prints: HI!
```



### 2.15.2 A Char Is Just a Number

Internally, a `char` is nothing more than a number — its **ASCII value**, in the range 0–255. There is therefore nothing stopping the same values from being printed as integers with `%i`:

```c
printf("%i %i %i\n", c1, c2, c3);   // prints: 72 73 33
```

`'H'` is **72**, `'I'` is **73**, and `'!'` is **33**. This demonstrates the equivalence between characters and numbers: storing `"HI!"` is really storing three patterns of 8 bits whose decimal values are 72, 73, and 33.

No explicit cast is required. `printf` is smart enough that when a `char` is supplied to a `%i` placeholder, it is treated as an integer automatically — an **implicit cast** rather than an explicit `(int)`.



### 2.15.3 A String Is an Array of Characters

A string holds a whole word or phrase — multiple characters. Declaring a string lays it out in memory exactly like the three separate chars above:

```c
string s = "HI!";
printf("%s\n", s);   // prints: HI!
```

Because a string is laid out as a contiguous sequence of characters, it can be indexed with **square-bracket notation**, just like any array:

```c
printf("%c%c%c\n", s[0], s[1], s[2]);   // prints: HI!
printf("%i %i %i\n", s[0], s[1], s[2]); // prints: 72 73 33
```

`s[0]` is `'H'`, `s[1]` is `'I'`, and `s[2]` is `'!'`. A **string is essentially an array of characters**.

> Note: `string` is a CS50 training-wheel type, so `#include <cs50.h>` is required for the compiler to recognize it.



### 2.15.4 The Null Terminator

If the location one byte *past* the end of `"HI!"` is examined, its value turns out to be **0**:

```c
printf("%i\n", s[3]);   // prints: 0
```

This is by design. By convention, the compiler automatically **terminates** every double-quoted string with a byte of all-zero bits. Written as the escape sequence **`\0`** (analogous to `\n`), this means literally **eight zero bits**, *not* the digit `0` that would be shown on screen. This special byte is the **null character**, or **NUL** for short — ASCII value 0 — and it marks the **end of a string**.

Consequently, `"HI!"` — though it shows three visible characters — actually occupies **4 bytes** of memory:

```
 s[0]   s[1]   s[2]   s[3]
+------+------+------+------+
|  'H' |  'I' |  '!' |  \0  |
|  72  |  73  |  33  |   0  |
+------+------+------+------+
```

Even an **empty string** `""` takes up **1 byte** — just the `\0`. This is why a string's size in memory depends on its length, **plus one**.



### 2.15.5 Why the Terminator Is Needed

Functions such as `printf` (with `%s`) and string-length code are deliberately "stupid": they scan character by character and need to know where to **stop**. The NUL byte provides that stopping point.

Consider two strings in memory at once:

```c
string s = "HI!";
string t = "BYE";
```

`"BYE"` occupies 4 bytes (`B`, `Y`, `E`, `\0`) and `"HI!"` occupies 4 bytes (`H`, `I`, `!`, `\0`). Without the terminating NUL, a function printing `s` would not know where `s` ends and would bleed straight into `t`. The choice of all-zero bits is arbitrary but agreed upon by convention.

> Distinguish the string terminator `\0` / NUL from the later **`NULL`** pointer — they are different things; the details come later.



### 2.15.6 Arrays of Strings

Since strings can themselves be stored in arrays, an **array of strings** can be declared:

```c
string words[2];
words[0] = "HI!";
words[1] = "BYE";
```

Because each element is *itself* an array of characters, **two sets of square brackets** can be used — the first index selects the word, the second selects a character within that word (counting from 0):

```c
// First word, character by character:
printf("%c%c%c\n", words[0][0], words[0][1], words[0][2]);     // HI!
// Second word, character by character:
printf("%c%c%c%c\n", words[1][0], words[1][1], words[1][2], words[1][3]); // BYE
```

`words[0][0]` is the first character of the first word; `words[1][2]` is the third character of the second word. This is simply the same indexing idea composed twice.

> An array can also be **statically initialized** with curly braces when all values are known up front — e.g. `int n[] = {72, 73, 33};` — in which case the size may be omitted, since the compiler can count the elements.

## 2.16 String Length and Libraries



### 2.16.1 Computing Length Manually

Because a string is an array ending in the null character, its length can be found by walking the array until the NUL is reached:

```c
int n = 0;
while (name[n] != '\0')
{
    n++;
}
printf("%i\n", n);
```

**Tracing the loop** for `name = "DAVID"`: `name[0]` is `'D'` (not `\0`, so `n` becomes 1), `name[1]` is `'A'` (`n` becomes 2), then `'V'`, `'I'`, `'D'` — incrementing `n` to 5. At `name[5]` the **null character** is reached, the condition is false, and the loop stops. Since the NUL itself is **not counted**, `n` holds 5.

- `"DAVID"` → **5**
- `"KELLY"` → **5**
- empty input (`""`) → **0** — literally true, because `get_string` still places a `\0` at the start, so the string occupies 1 byte but contains 0 visible characters.



### 2.16.2 strlen

This wheel does not need to be reinvented. Beyond `stdio.h`, `cs50.h`, and `math.h` (used in Problem Set 1), other libraries exist — including the **string library**, **`string.h`**. It provides **`strlen`** (string length), documented in the CS50 manual at **manual.cs50.io**. Someone wrote essentially the loop above decades ago and packaged it as a reusable function.

```c
#include <string.h>
...
int n = strlen(name);
printf("%i\n", n);
```

For `"DAVID"` and `"KELLY"` this prints 5; for empty input it prints 0 — matching the manual version.

By good design, the intermediate variable can be dropped, since functions can be **nested**:

```c
printf("%i\n", strlen(name));
```



### 2.16.3 Loop Efficiency

Consider a program that prints a string one character at a time:

```c
string s = get_string("Input:  ");
printf("Output: ");
for (int i = 0; i < strlen(s); i++)
{
    printf("%c", s[i]);
}
printf("\n");
```

This is **correct but not well designed**. Recall how a `for` loop runs: initialization happens once, then on *every* pass the Boolean condition is re-evaluated before the body and update. That means **`strlen(s)` is called on every single iteration** — even though the length of `s` never changes. Repeatedly re-measuring the same string is wasted work.

**Fix 1 — a separate variable**, so the length is computed only once:

```c
int n = strlen(s);
for (int i = 0; i < n; i++)
{
    printf("%c", s[i]);
}
```

**Fix 2 — declare both variables in the init clause** of the `for` loop:

```c
for (int i = 0, n = strlen(s); i < n; i++)
{
    printf("%c", s[i]);
}
```

Here `i` is set to 0 and `n` to `strlen(s)` once, then reused. The **constraint** is that all variables declared in the init clause must be the **same type** (both `int` here); if a different type were needed, the value would have to be pulled out into a separate declaration as in Fix 1.

> **Caveat:** Modern compilers are advanced enough that they may notice the repeated `strlen` call and optimize it away automatically. Even so, this is a class of inefficiency worth spotting by eye. In a loop running millions of times, those wasted microseconds add up to real compute, time, and money.



### 2.16.4 A Note on Spaces

A space is simply **ASCII 32** — an ordinary character that happens to be invisible. It is counted by `strlen` and treated like any other character; there is no special accounting for it. **Only the null character is special**, because functions like `printf` and `strlen` look for it to determine where a string ends.

## 2.17 Uppercasing and the ctype Library

A program that uppercases a string typed by the user illustrates a progression from "good" to "better" to "best."



### 2.17.1 Manual Approach Using ASCII Math (uppercase.c)

The headers `cs50.h`, `stdio.h`, and `string.h` are included. After obtaining input with `get_string`, the program iterates over the string and forces each lowercase character to uppercase.

Because characters **are** their ASCII values, chars can be compared like ints. The ASCII chart from Week 0 assigns numbers to both uppercase and lowercase letters, all contiguous (lowercase `'a'` is **97**, uppercase `'A'` is **65**). A character is lowercase when it falls between `'a'` and `'z'` inclusive.

Every lowercase letter sits exactly **32** greater than its uppercase counterpart, so **subtracting 32** converts lowercase to uppercase. Non-lowercase characters are printed unchanged.

```c
for (int i = 0, n = strlen(s); i < n; i++)
{
    // if s[i] is lowercase, force it to uppercase
    if (s[i] >= 'a' && s[i] <= 'z')
    {
        printf("%c", s[i] - 32);
    }
    // else if not lowercase, print unchanged
    else
    {
        printf("%c", s[i]);
    }
}
```

| Character | ASCII value | Result of `- 32` |
|-----------|-------------|------------------|
| `'a'` | 97 | 65 → `'A'` |
| `'A'` | 65 | (left unchanged) |

The literal `32` is safe to hard-code because ASCII is a fixed standard that will not change.



### 2.17.2 Better: The ctype.h Library

Writing this much code for something as commonplace as toggling case (a one-click operation in Word or Google Docs) is tedious. The **ctype.h** library provides character functions that solve this directly.

- **`islower(c)`** — returns true if the character is lowercase.
- **`toupper(c)`** — returns the uppercase version of a character, passing any non-lowercase character through unchanged.

`islower` already encapsulates the "is it between `'a'` and `'z'`" check, so the manual comparison is unnecessary. Because `toupper` leaves non-lowercase characters unchanged, the entire `if/else` collapses to a single line:

```c
#include <ctype.h>
...
printf("%c", toupper(s[i]));
```

The function itself decides whether to convert: lowercase letters are changed first, everything else is returned as-is.



### 2.17.3 Best Practice

When coding becomes tedious or a problem has surely been solved before, a library function — from CS50 or from the standard library — likely already exists.



### 2.17.4 Note on Linking

Standard libraries such as **ctype.h** and **string.h** link automatically; no `-l` flag is required. Only non-standard libraries like CS50's (`-lcs50`, the training wheels for the first few weeks) must be linked explicitly. `make` is configured to handle all of this automatically.

## 2.18 The main Function: argc and argv

`main` is special: in C (as in C++, Java, and certain other languages) it is the function called automatically once a program is compiled and run. Two forms are supported, and only two.



### 2.18.1 main(void)

```c
int main(void)
```

The `void` in parentheses means `main` takes **no command-line arguments**. This is why every program written so far — `hello`, `scores`, and the rest — only ever accepted input through `get_string`, never through words typed after the program's name. This differs from commands like `make`, `code`, and `cd`, which do accept words after their names at the prompt.



### 2.18.2 main(int argc, string argv[])

```c
int main(int argc, string argv[])
```

The other supported form takes exactly two arguments:

- **argc** (*argument count*) — the number of words typed at the prompt, **including the program's own name**.
- **argv** (*argument vector*) — an **array of strings** holding the actual words.

Now that arrays are available, a program can receive 0, 1, 2, 3, or more words from the user at the prompt.



### 2.18.3 greet.c Example

Converting `greet.c` from `get_string` input to command-line arguments requires only changing the prototype from `void` to `int argc, string argv[]` and printing `argv[1]`:

```c
#include <cs50.h>
#include <stdio.h>

int main(int argc, string argv[])
{
    printf("hello, %s\n", argv[1]);
}
```

```bash
$ ./greet David
hello, David
$ ./greet Kelly
hello, Kelly
```

The `string` type still requires `#include <cs50.h>`, since `string` is defined there. (`string` is a term of art, not a C keyword; Week 4 reveals what it truly is.)



### 2.18.4 argv[0] Is Always the Program's Name

Running the program with no further words shows that location 0 of `argv` automatically holds the program's own name:

```bash
$ ./greet
hello, greet
```

This is useful for self-referential output — thanking the user for running the program, or displaying documentation that reflects whatever the file has been named or renamed to. `argv[0]` always contains the program's name.



### 2.18.5 Guarding Against Missing Arguments with argc

Running `./greet` with no argument prints `hello, (null)` — `printf` knows no real word is present. (This `NULL` is **not** the same as the string terminator `NUL`; the distinction comes later.)

The `argc` count guards against this:

```c
if (argc == 2)
{
    printf("hello, %s\n", argv[1]);
}
else
{
    printf("hello, world\n");
}
```

| Command | argc | Behavior |
|--------------------|------|---------------------|
| `./greet` | 1 | `hello, world` |
| `./greet David` | 2 | `hello, David` |
| `./greet David Malan` | 3 | rejected → `hello, world` |

`./greet David Malan` makes `argc == 3` (program name plus two words), so the `argc == 2` check rejects it. The first word is always the program's name.



### 2.18.6 Why This Form Exists

This is simply the definition of C: `main` may take no arguments (`void`) or exactly two (`argc` and `argv`). The compiler and operating system fill those two variables automatically when arguments are provided. The variable names may be renamed, but the **types and count cannot change** — three or four arguments are not permitted.

## 2.19 Command-Line Arguments in Depth

The previous section introduced the two forms of `main` and used `argv[0]` / `argv[1]` in `greet.c`. This section deepens the treatment of how `argc` and `argv` actually work.



### 2.19.1 Runtime vs. In-Program Prompts

Until now, user input has been collected with **in-program prompts** — `get_int`, `get_float`, `get_string` — *while* the program is running. Command-line arguments are a different channel: they let the user provide data **at the moment the program is run**, before it starts executing. Collecting them requires the second form of `main`:

```c
int main(int argc, string argv[])
{
    ...
}
```

The names `argc` and `argv` are **conventions** (argument count, argument vector); they could be named anything, though using the conventional names is standard practice.



### 2.19.2 `argc` — Argument Count

`argc` is an integer holding **how many command-line arguments were typed**, including the program's own name. Arguments are separated by **whitespace** (spaces, tabs).

| Command | `argc` |
|---------|--------|
| `./greedy` | 1 |
| `./greedy 1024 cs50` | 3 |

The program name itself always counts as one argument, so a program run with no extra words still has `argc` of 1.



### 2.19.3 `argv` — Argument Vector

`argv` ("vector" is just another word for array) is an **array of strings**, one per typed argument:

- The **first** element is always `argv[0]` — the program's name.
- The **last** element is always `argv[argc - 1]`.

The `argc - 1` formula follows directly from zero-based indexing: an array of `argc` elements has valid indices `0` through `argc - 1`. For `./greedy 1024 cs50` (`argc` = 3), the elements are `argv[0]`, `argv[1]`, `argv[2]` — there is **no** `argv[3]`.

| Index | Contents (`./greedy 1024 cs50`) |
|-------|--------------------------------|
| `argv[0]` | `"./greedy"` |
| `argv[1]` | `"1024"` |
| `argv[2]` | `"cs50"` |
| `argv[3]` | out of bounds → `???` (likely **segmentation fault**) |

Accessing `argv[3]` here oversteps the array. As with any C array, the compiler permits it, but at run time it reads memory the program was not given — and with `argv` especially, this very often causes a **segmentation fault**.



### 2.19.4 Everything in `argv` Is a String

This is the most important and easily-missed point: **every element of `argv` is a string**, even when it looks like a number.

For `./greedy 1024 cs50`, `argv[1]` is the **string** `"1024"` — the four characters `'1'`, `'0'`, `'2'`, `'4'` — **not** the integer `1024`. Consequently, arithmetic on it does not behave numerically:

```c
int x = argv[1] - 24;   // does NOT give 1000 — argv[1] is a string, not an int
```

To use a command-line argument as a number, it must first be **converted** from a string to an integer with a conversion function (such as `atoi`, documented in the CS50 reference and introduced with a later problem set). This string-vs-number distinction is a common source of bugs when first working with command-line arguments.

### 2.20.1 Numeric Codes in Computing

## 2.21 Cryptography: The Caesar Cipher

**Cryptography** is the art of scrambling information for secure communication — important for passwords, credit card numbers, and personal messages.



### 2.21.1 Mental Model

```
plaintext  →  [ cipher ]  →  ciphertext
```

- **Plaintext** — the human-readable message.
- **Cipher** — an algorithm (the "black box") that **encrypts** the information.
- **Ciphertext** — the scrambled output.

Plaintext cannot simply be run through an algorithm to produce ciphertext that the recipient can reverse; a **secret key** must be shared in advance. Without an agreed-upon secret, the recipient could not reverse the scrambling.



### 2.21.2 The Caesar Cipher

Each letter is shifted by a fixed key. With **key = 1**, every letter advances one place, wrapping around at the end of the alphabet:

```
A → B,  B → C,  ...,  Z → A
```

Encrypting `"HI"` with key 1 yields ciphertext `"IJ"`. **Decryption** uses the same key but subtracts, inverting +1 to −1 (`B → A`, `C → B`, `A → Z`). In this simple cipher, punctuation and other non-letters are typically left unchanged.



### 2.21.3 ROT13 and ROT26

**ROT13** (key 13) has special significance in computing circles. It is used online — for example on Reddit — to hide spoilers, requiring a small effort (a click or a line of code) to reveal the message. With key 13, `"HI"` becomes `"UV"`, since `U` and `V` are 13 places away from `H` and `I` respectively.

**ROT26** is useless: rotating 26 places is a full wrap-around, mapping every letter back to itself (`A → A`, `B → B`).



### 2.21.4 Security and Weakness

A Caesar key can only be 1 through 25, so it is trivially **brute-forced**. From an adversary's perspective — a teacher intercepting a passed note — one simply tries every key (1, 2, 3, … 25) until readable plaintext appears. The cipher is therefore **not secure**.

Historically it mattered only because Caesar's contemporaries did not know the scheme — **security through obscurity**. Today, writing code to brute-force it is trivial, so modern algorithms are far more sophisticated than simple letter rotations. Decryption, however, remains fundamentally just **reversing the process**.

This week's problem set has students implement encryption in code, exploring the algorithms by which information can be enciphered.

### 2.22.1 Heuristics for Good Code Design

This section recaps a CS50 discussion section (a small-group practice session led by a preceptor, distinct from the main lecture) in which good and bad code-design choices are compared side by side. Code design is one of the axes on which problem sets are graded, alongside correctness and style. The slide deck frames the question as two top-level criteria, each expanded into its own heuristics below: **is the code efficient?** and **is the code easy to maintain?**

#### Is Your Code Efficient?

- **Efficiency** — code should avoid unnecessary repetition and redundant operations.
- **Don't Repeat Yourself (DRY)** — copying and pasting code, even with slight modifications, should be avoided. Repeated, copy-pasted code is a sign that a **loop** or an **abstraction** (a function) should be used instead to hide the repetition.
- **Avoiding redundant computation** — if a value is computed multiple times without changing, it should be computed once and stored in a variable, rather than recomputed on each use.

#### Is Your Code Easy to Maintain?

- **Maintainability** — code should scale gracefully as the size of the problem grows.
- **Prototyping functions above `main`** and **avoiding magic numbers** (both detailed below) are the slide deck's two concrete maintainability practices.

#### Prototyping Functions Above `main`

A **function prototype** is the first line that declares a function — typically a helper function that assists in solving the problem — placed *above* `main`. This convention makes code easier for other humans to read, and keeps `main`, the entry point and overview of the program's logic, at the top of the file.

#### Avoiding Magic Numbers

A **magic number** is a hard-coded numeric literal whose meaning is not obvious from context and whose value will not change over the course of the program's execution. Such numbers should instead be defined as named **constants**.

> The alphabet has 26 letters. Since this fact never changes during a program's run, the literal `26` should be defined as a constant rather than typed directly wherever it is needed — relevant to this week's problem set tasks involving letters of the alphabet.

#### Comparison Game: Option A vs. Option B

Four paired code snippets are used to illustrate the heuristics above. The slides present each pair under the same prompt — "Which is better and why?" — before identifying and justifying the better-designed option.

##### Comparison 1 — Recomputing String Length vs. Caching It

```c
// Option A
for (int i = 0; i < strlen(s); i++)
{
    printf("%c\n", s[i]);
}
```

```c
// Option B
int n = strlen(s);
for (int i = 0; i < n; i++)
{
    printf("%c\n", s[i]);
}
```

**Option B is better designed.** In Option A, `strlen(s)` is recomputed on every iteration of the loop, which is unnecessary computation since the string's length is not expected to change while the loop runs. Option B computes the length once, stores it in `n`, and reuses that value, avoiding redundant work.

##### Comparison 2 — Magic Number vs. Named Constant

```c
// Option A
const int MONTHS_IN_YEAR = 12;
// ... uses MONTHS_IN_YEAR throughout
```

```c
// Option B
// ... uses the literal 12 directly throughout
```

**Option A is better designed.** The number of months in a year does not change over the course of a program's execution, so defining it as a named constant (`MONTHS_IN_YEAR`) is preferable to scattering the literal `12` through the code, where a reader might not immediately understand what the number represents. This is a direct application of avoiding magic numbers.

##### Comparison 3 — Redundant `else` After a Return

```c
// Option A
if (expression)
{
    return 0; // success
}
return 1;
```

```c
// Option B
if (expression)
{
    return 0; // success
}
else
{
    return 1;
}
```

**Option A is better designed.** Both options check the same condition. In Option B, the `else` branch is redundant: if the `if` branch is not taken, execution falls through to `return 1` regardless of whether an explicit `else` wraps it, because the `if` branch already returns — exiting the function — when its condition is true. There is no scenario in which the `else` is needed to prevent `return 1` from running unintentionally. Option A achieves identical behavior with one fewer line and no unnecessary branching structure.

> An `else` block is redundant whenever the code inside it would run regardless of whether the `else` keyword were present — i.e., when the preceding `if` block unconditionally exits (for example, via `return`). A useful test: "would the code in the `else` block run in either case, with or without the `else`?" If yes, the `else` is unnecessary.

##### Comparison 4 — Guard Clause vs. Wrapping Code in `if`

```c
// Option A
bool success = ...;
if (success)
{
    // large block of code
}
else
{
    // exit case, e.g. stop the program
}
```

```c
// Option B
bool success = ...;
if (!success)
{
    // exit case, e.g. stop the program
}
// large block of code
```

**Option B is better designed.** A boolean variable such as `success` stores `true` or `false`. In Option A, the large body of code is nested inside the `if (success)` branch, with the exit/failure logic in the `else`. In Option B, the failure condition is checked first as a **guard clause** at the top of the function: if the check fails, the program exits or returns immediately, before any of the large block of code runs. This is more succinct (it avoids an extra `else` case) and arguably easier to follow.

> Guard clauses act as guardrails placed at the beginning of a function, before the bulk of the logic, making the control flow easier to follow.

| | Option A | Option B | Better choice | Why |
|---|---|---|---|---|
| 1. String length | Recomputes `strlen(s)` every iteration | Caches length in `n` before the loop | B | Avoids redundant computation; the length does not change during the loop |
| 2. Months in a year | Defines `MONTHS_IN_YEAR` as a named constant | Uses the literal `12` directly throughout | A | Avoids a magic number; the constant communicates meaning and the value never changes |
| 3. Return after `if` | Omits the `else`; falls through to `return 1` | Wraps the second `return` in an `else` | A | The `if` branch already exits via `return`, making the `else` redundant |
| 4. Success handling | Nests the large code block inside `if (success)`, with failure in `else` | Uses a guard clause (`if (!success) { exit }`) before the large code block | B | The guard clause front-loads the failure case, avoiding nested `else` and easing readability |

#### When Is an `else` Not Redundant?

Whether an `else` is necessary depends on the code being written. A useful test is to ask whether the code inside the `else` block would run in *either* case — whether the `else` is present or not. If the code runs the same way regardless, the `else` is redundant and can be removed. `else` (or `else if`) becomes necessary chiefly when there are more than two possible outcomes/branches to distinguish.

### 2.22.2 Arrays: Motivation and Declaration

#### Motivating Example: Storing a Leaderboard

A motivating example for arrays is storing the top five scores of players in a video game such as Pac-Man, as on a leaderboard.

Without arrays, the only available approach is declaring five separate integer variables: `score1`, `score2`, `score3`, `score4`, `score5`. This approach has drawbacks: it requires keeping track of many separate variables, and it does not scale well as the number of scores grows.

An **array** provides a way of categorizing and storing a fixed collection of related items together under a single name, rather than as separate variables — for example, one array holding all five Pac-Man scores.

#### Declaring an Array

```c
int scores[5];
```

This declaration is composed of three parts:

This declares an empty array of five integers; it does not yet contain any specific values.

### 2.22.3 Strings as Arrays of Characters

#### 2.22.3.1 Strings as Arrays of Characters

A **string** is not a native keyword/type in C itself; the `string` type used in CS50 code is a training-wheel construct provided by the CS50 library, layered on top of what a string actually is at a lower level (covered in more depth elsewhere).

A string can be understood as an **array of characters**. For example, given:
```c
string name = "Emma";
```
`name` can be thought of as an array containing the characters `E`, `m`, `m`, `a`, followed by a special terminating character, `\0` (written as backslash-zero):

```text
name

E   m   m   a   \0

0   1   2   3   4
```

> The `\0` (pronounced "NUL") character marks the **end of the string** in memory. When a string is stored in memory, encountering `\0` signals exactly where the string ends.

Because a string is just an array of characters, the same square-bracket indexing notation used for arrays can be used to access individual characters:
```c
name[0]   // 'E' — the first character
name[1]   // 'm' — the second character
```
Indexing is 0-based, consistent with array indexing generally.

#### 2.22.3.2 ASCII Representation of Characters

Each character in a string can also be represented as a number via the **ASCII table**. For example, capital `A` corresponds to ASCII value `65`; adding `1` to `65` and reinterpreting the result as a character yields `B`.

> **Reference (ASCII letter ranges):** Uppercase `A`–`Z` = `65`–`90`; lowercase `a`–`z` = `97`–`122`.

Because each character is stored as its underlying ASCII number, the string `"Emma"` is really stored as the numeric sequence below, with `\0` corresponding to ASCII `0`:

```text
name

69   109   109   97   \0(0)

0    1     2     3    4
```

Characters correspond to ASCII numeric values, so characters can be compared and operated on numerically — e.g. lowercase `b` = `98`, lowercase `a` = `97`; since `98 > 97`, `b` comes after `a`. This numeric ordering is the foundation for comparing characters in later exercises (see §2.22.3.7).

#### 2.22.3.3 print.c — Printing a String Character by Character

> **Task:** Write a program called `print.c` that takes a string as input and prints it character-by-character.
>
> ```text
> $ ./print
> Input: Hello
> Hello
> ```

Headers needed: `cs50.h`, `stdio.h`, `string.h`. Construction reasoning: input is requested via `get_string`; printing repeats once per character, so a loop is needed; a `for` loop is chosen; the loop bound is not known ahead of time, so `strlen(text)` is used as the bound; `%c` prints a single character.

```c
#include <cs50.h>
#include <stdio.h>
#include <string.h>

int main(void)
{
    string text = get_string("Text: ");

    for (int i = 0; i < strlen(text); i++)
    {
        printf("%c", text[i]);
    }
}
```

**Design review — efficiency improvement.** The condition `i < strlen(text)` recomputes `strlen(text)` on every iteration, even though the text's length never changes — this is wasted computation. The fix caches the length in a variable before the loop:

```c
int length = strlen(text);
for (int i = 0; i < length; i++)
{
    printf("%c\n", text[i]);
}
```

> Takeaway: avoid calling a function repeatedly inside a loop condition when its result will not change across iterations; cache the result in a variable instead.

Stylistic note: placing `\n` inside the loop body puts every character on its own line; placing `\n` after the loop instead prints all characters together followed by a single trailing newline — a stylistic choice, not a correctness issue.

Sample run (with the per-line `\n` version, input `Kelly`):
```text
$ ./print
Text: Kelly
K
e
l
l
y
```

#### 2.22.3.4 reverse.c — Reversing a String

> **Task:** Write a program, `reverse.c`, that takes a string as input and reverses it.
>
> ```text
> $ ./reverse
> Text: This is CS50.
> Reverse: .05SC si sihT
> ```

Built by copying `print.c` and modifying the loop: instead of starting at the beginning, `i` starts at the end of the string and decrements, with loop condition `i >= 0`.

**Bug encountered.** The loop was initially written to start `i` at `strlen(text)`. For input `cat` (length 3), `strlen(text)` returns `3`, but because of zero-indexing, index `3` is *not* a valid character position — it is the null terminator `\0`. The last real character sits at index `length - 1`. The fix starts the loop at `strlen(text) - 1`:

```c
int length = strlen(text);
for (int i = length - 1; i >= 0; i--)
{
    printf("%c", text[i]);
}
```

> Caution: `strlen(text)` gives the *count* of characters, but valid indices run `0` to `strlen(text) - 1`. Index `strlen(text)` itself is the null terminator.

```text
$ ./reverse
Text: Kelly
ylleK
```

#### 2.22.3.7 alphabetical.c — Checking Alphabetical Order via ASCII Comparison

> **Task:** Write a program, `alphabetical.c`, that prints `Yes` if a lowercase string's characters are in alphabetical order, and `No` if they are not.
>
> ```text
> $ ./alphabetical
> Input: abcdefg
> Yes
> $ ./alphabetical
> Input: gfedcba
> No
> ```

Built from the `print.c`/`reverse.c` base, comparing each character to its successor using the ASCII ordering introduced in §2.22.3.2.

An initial condition was drafted as `text[i] < text[i + 1]` (true when the pair is in order). This phrasing was discarded: "if in order, do nothing" leaves no natural action for the true branch. The logic was restructured to check for the *violation* instead — if a character is ever found greater than the one following it, the string is immediately known not to be in alphabetical order:

```c
if (text[i] > text[i + 1])
{
    printf("No\n");
    return 0;
}
```

If the loop completes without ever returning, every adjacent pair was in order, so `Yes` is printed after the loop.

**Bug encountered.** Test input `abc` was incorrectly reported as *not* alphabetical. Diagnosed by printing the ASCII values of `text[i]` and `text[i + 1]` on each iteration. Tracing `abc`:

| Comparison | Result |
|---|---|
| `'a'` vs `'b'` | in order |
| `'b'` vs `'c'` | in order |
| `'c'` vs `text[3]` (`\0`, ASCII `0`) | `0 < ASCII('c')`, so `text[i] > text[i+1]` evaluates **true** |

The last comparison is illegitimate: at that point `i` has advanced past the last real character, so `text[i + 1]` is the null terminator rather than a real character. Since `0` is less than the ASCII value of `'c'`, the violation condition fires incorrectly, causing the program to print `No` for a string that is in fact alphabetical.

**Root cause.** The loop compares one index pair too many — it allows `i` to reach a position where `text[i + 1]` is the null terminator `\0` rather than a genuine character.

> Caution: when comparing `text[i]` to `text[i + 1]` in a loop over a string of length `n`, the loop must not let `i` reach `n - 1`, since doing so compares against `text[n]`, the null terminator. The loop bound needs adjusting (e.g., to `i < length - 1`) to avoid this.

### 2.22.4 Command-Line Arguments Revisited

Building on the argc/argv mechanics introduced earlier (see 2.18–2.19), the section revisits command-line arguments with three worked examples and two practice exercises.

#### 2.22.4.1 argc/argv Worked Examples

Three familiar invocations illustrate how `argc` and `argv` are populated at runtime.

**`make mario`**

| Variable | Value |
|---|---|
| `argv[0]` | `"make"` |
| `argv[1]` | `"mario"` |
| `argc` | `2` |

**`./caesar 13`**

| Variable | Value |
|---|---|
| `argv[0]` | `"./caesar"` |
| `argv[1]` | `"13"` |
| `argc` | `2` |

> `argv[1]` holds the string `"13"`, not the integer `13`. Every element of `argv` is a string, so converting it to a number requires explicit conversion (e.g., `atoi`, per 2.19.4).

**`./initials Kelly Ding`**

A hypothetical program `initials` takes a name as command-line arguments and prints the corresponding initials (`Kelly Ding` → `KD`).

| Variable | Value |
|---|---|
| `argv[0]` | `"./initials"` |
| `argv[1]` | `"Kelly"` |
| `argv[2]` | `"Ding"` |
| `argc` | `3` |

`argv[1][0]` is `'K'`; `argv[2][0]` is `'D'`.

> `argc` counts from 1, like any count (e.g., an array's length), but indices into `argv` count from 0. For `./initials Kelly Ding`, `argc` is `3`, while the valid indices into `argv` are `0`, `1`, and `2`.

#### 2.22.4.2 initials.c (via argv)

Unlike the `get_string`-based `initials.c` in 2.22.3, this version accepts the name directly as command-line arguments. Names are assumed to be separated by a single space, with each word arriving as a separate element of `argv`.

```c
#include <cs50.h>
#include <stdio.h>

int main(int argc, string argv[])
{
    if (argc < 2)
    {
        printf("Usage: ./initials First Last\n");
        return 1;
    }

    for (int i = 1; i < argc; i++)
    {
        printf("%c", argv[i][0]);
    }

    printf("\n");
}
```

Sample runs:

```text
$ ./initials Kelly Ding
KD
```

```text
$ ./initials
Usage: ./initials First Last
```

```text
$ ./initials David Malan
DM
```

> Design decision: requiring `argc < 2` (rather than `argc != 3`) accepts any number of name parts, including middle names. Checking for `argc != 3` would incorrectly reject a valid four-argument invocation such as `./initials David J. Malan`.

#### 2.22.4.3 addition.c

**Task:** Write a program, `addition.c`, that adds two numbers provided as command-line arguments.

```text
$ ./addition 2 8
2 + 8 = 10
$ ./addition 2
Usage: ./addition x y
```

Since `argv` elements are strings (2.19.4), each numeric argument must be converted with `atoi` — the only such conversion function introduced so far — before being summed.

```c
#include <stdio.h>
#include <stdlib.h>

int main(int argc, char *argv[])
{
    if (argc != 3)
    {
        printf("Usage: ./addition x y\n");
        return 1;
    }

    int x = atoi(argv[1]);
    int y = atoi(argv[2]);

    printf("%i + %i = %i\n", x, y, x + y);
}
```
