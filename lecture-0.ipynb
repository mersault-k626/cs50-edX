# CS50 Lecture 0 Key Notes


## What is Computer Science?
- It is the study of information and problem-solving.
- **Computational Thinking**: Application of computer science ideas to real-world problems.
- Problem solving involves turning **Input** into **Output** through an **Algorithm** (a "black box" of step-by-step instructions).

## Information Representation
- Computers use the **Binary System** (Base 2) consisting of 0s and 1s, driven by electricity (on/off states).
- **Bit**: A single binary digit (0 or 1).
- **Byte**: 8 bits. Used as standard units of measure.
- **Numbers**: Binary translates directly to decimal digits using powers of 2.
- **Text**: 
  - **ASCII**: Standardized mapping of numbers to characters (e.g., 'A' is 65).
  - **Unicode**: A superset of ASCII using more bits (16, 24, or 32 bits) to represent all human languages and symbols, including emojis.
- **Colors & Images**: Uses RGB (Red, Green, Blue). Each color intensity is stored as a byte (0-255). A single pixel uses 3 bytes (24 bits). Images are collections of colored pixels.
- **Videos**: Sequential images displayed quickly.
- **Audio**: Sound notes mapped to integers representing frequency, duration, and volume.

## Algorithms
- Algorithms are step-by-step instructions.
- Efficiency matters: e.g., Finding a name in a phone book:
  - **Linear Search**: Searching page by page (slowest).
  - **Divide and Conquer**: Tearing the phone book in half to discard half the problem repeatedly (logarithmic time, much faster).

## Pseudocode & Programming Concepts
- **Functions**: Verbs or actions that perform a specific task.
- **Conditionals**: Forks in the road (e.g., `If/Else`), driven by Boolean expressions.
- **Boolean Expressions**: Questions evaluating to true/false or yes/no.
- **Loops**: Inducing cyclical behavior to avoid redundant code.
- **Variables**: Storing values in memory.
- **Abstraction**: Hiding lower-level implementation details so programmers can focus on solving complex problems.

## Scratch
- A graphical programming language using puzzle-piece blocks to represent lower-level code concepts.
- **Sprites**: Characters/objects on the screen.
- Showcases loops, functions, variables, side effects (visual changes), and return values (data).
