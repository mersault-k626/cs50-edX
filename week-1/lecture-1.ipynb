## 1.1 From Scratch to C

The building blocks introduced in Scratch — functions, conditionals, loops, variables, abstraction — are the same ones used in every programming language. What changes week to week is the syntax, not the underlying ideas.

In Scratch, a "Hello, world" program was two puzzle pieces:
```
when green flag clicked
say [Hello, world!]
```

In C, the same program is:
```c
#include <stdio.h>

int main(void)
{
    printf("Hello, world\n");
}
```

The structure maps directly: the `when green flag clicked` block → the boilerplate scaffolding; the `say` block → `printf`; the text in the oval → the string inside the parentheses.

### 1.1.1 What each line means

```c
#include <stdio.h>
```
Pulls in an external file full of pre-written tools. `stdio.h` stands for "standard input/output" — it gives your program access to `printf`. Without this line, the compiler doesn't know what `printf` is. Think of it like `import` in Python or `library()` in R — prerequisites loaded before the actual program runs.

```c
int main(void)
```
Every C program must have a function called `main` — it's where the program starts running. `int` means the function will hand back a number when it finishes (by convention, `0` means success). `void` inside the parentheses means it takes no input.

```c
{
}
```
Curly braces define a **block** — everything inside `{}` belongs to `main`. The opening `{` marks the start; the closing `}` marks the end.

```c
    printf("Hello, world\n");
```
`printf` ("print formatted") prints text to the terminal. Breaking down each part:

- `""` — the quotes wrap the text you want to print; without them C doesn't know where the text starts and ends
- `\n` — a newline character (like pressing Enter), so the cursor moves to the next line after printing; optional but without it the shell prompt appears on the same line as your output. The `\` signals that the next character has a special meaning — these are called **escape sequences**:

| Sequence | Effect |
|----------|--------|
| `\n` | new line (move cursor to next line) |
| `\t` | horizontal tab |
| `\\` | literal backslash |
| `\"` | literal double-quote inside a string |

- `;` — mandatory; every statement in C must end with a semicolon, like a period ending a sentence; missing it causes a compiler error

### Source Code, Machine Code, and the Compiler

Computers only execute **machine code** — zeros and ones. Humans write **source code** — text in a language like C. A **compiler** is a program that translates one into the other.

```
Source code  →  [ Compiler ]  →  Machine code
   (hello.c)                       (hello)
```

- **Source code**: what the programmer writes; human-readable with training.
- **Machine code**: what the CPU executes; zeros and ones.
- **Compiler**: a piece of software that bridges the two. In CS50, `make` is used as the command that invokes the compiler (the actual compiler underneath is called `clang`).

The workflow in CS50's environment:
```bash
code hello.c      # open/create the source file in VS Code
make hello        # compile hello.c → produces executable called hello
./hello           # run the compiled program
```

If no output appears after `make`, the compilation succeeded. Output (error messages) means something is wrong in the source code.

## 1.3 C Syntax Fundamentals

### 1.3.1 `printf` and Arguments

`printf` ("print formatted") is C's primary output function. It takes a **format string** as its first argument, with optional additional arguments for placeholders:

```c
printf("Hello, world\n");           // one argument
printf("Hello, %s\n", name);        // two arguments: format + variable
printf("%i + %i = %i\n", x, y, z); // four arguments
```

The comma inside the quotes is English punctuation. The comma *outside* the quotes separates arguments to `printf`. These are different things occupying the same line.

---

### Escape Sequences

Certain characters inside a string have special meaning when preceded by a backslash:

| Sequence | Effect |
|----------|--------|
| `\n` | new line (move cursor to next line) |
| `\t` | horizontal tab |
| `\r` | carriage return (move cursor to start of line) |
| `\\` | literal backslash |
| `\"` | literal double-quote inside a string |
| `\'` | literal single-quote inside a char |

Without `\n`, the shell prompt appears on the same line as the output — aesthetically wrong even if technically functional.

---

### Semicolons and Structure

In C, every statement ends with a semicolon. Missing one produces a compiler error pointing to the line and character where it was expected:

```
hello.c:5:29: error: expected ';' after expression
```

The compiler error message gives: the filename, the line number, the character position, and a description. Reading the first error message (not the last) is usually the most productive approach.

---

### Header Files and Libraries

A **library** is code someone else wrote that can be used in your own programs. A **header file** (ending in `.h`) declares the functions a library provides so the compiler knows they exist.

To use a library, include its header at the top of the file:
```c
#include <stdio.h>    // standard input/output — gives access to printf, etc.
#include <cs50.h>     // CS50's own library — gives access to get_string, get_int, etc.
```

If `printf` is used without `#include <stdio.h>`, the compiler produces:
```
error: call to undeclared library function 'printf'
```

## 1.4 Data Types and Variables

C is a *statically typed* language — every variable must be declared with a type before use. The type tells the compiler how many bits to allocate and how to interpret those bits.

---

### 1.4.1 Primitive Types in C

| Type | Description | Size | Range / Precision |
|------|-------------|------|-------------------|
| `char` | single character | 8 bits (1 byte) | −128 to 127; maps to ASCII 0–127 |
| `int` | integer | 32 bits (4 bytes) | ~−2 billion to ~+2 billion |
| `unsigned int` | non-negative integer | 32 bits | 0 to ~4 billion |
| `long` | larger integer | 64 bits (8 bytes) | ~−9 quintillion to ~+9 quintillion |
| `float` | decimal number | 32 bits | ~6–7 significant digits of precision |
| `double` | higher-precision decimal | 64 bits | ~15–16 significant digits of precision |

**Types provided by `cs50.h` (not built into C):**

| Type | Description |

### `void` — a Type, Not a Data Type

`void` exists as a type but cannot be used to declare variables. It appears in two contexts:

- **Return type of a function**: `void meow(void)` means the function returns nothing.
- **Parameter list of a function**: `int main(void)` means the function takes no arguments.

`printf` is a `void` function — it prints to the screen as a side effect but returns nothing usable.

---

### `unsigned` Modifier

Adding `unsigned` before `int` (or other integer types) removes the ability to store negative numbers, and redirects those bits to double the positive range:

| Type | Range |
|------|-------|
| `int` (signed) | −2,147,483,648 to +2,147,483,647 |
| `unsigned int` | 0 to 4,294,967,295 |

Use `unsigned int` when values are known to never be negative and might exceed 2 billion.

### Declaring and Assigning Variables

```c
// Declaration + assignment (initialization) — most common form
string name = get_string("What's your name? ");
int counter = 0;
float price = 1.99;
char grade = 'A';       // single characters use single quotes
```

Multiple variables of the same type can be declared on one line:
```c
int height, width;          // declares two ints
float sqrt2, sqrt3, pi;     // declares three floats
```

**The `=` sign is the assignment operator** — it copies the value on the right into the variable on the left. It does not mean mathematical equality.

After a variable is declared, the type is never specified again. Re-specifying the type is either an error or creates a new shadowing variable:
```c
int n = 5;
int n = 10;   // BUG — re-declaring n; causes unpredictable behavior
n = 10;       // CORRECT — reassigning n (no type)
```

### Incrementing and Decrementing

```c
counter = counter + 1;  // verbose
counter += 1;           // shorthand: add 1
counter++;              // shortest: increment by exactly 1
counter--;              // decrement by exactly 1
counter *= 2;           // double it
counter /= 4;           // divide by 4
counter %= 6;           // replace with remainder when divided by 6
```

`++` and `--` only work for incrementing/decrementing by exactly 1. For any other amount, use `+= n` or `-= n`.

## 1.5 Conditionals

Conditional expressions allow programs to make decisions and take different **branches** in the road, depending on the values of variables or user input.

In C, every non-zero value is treated as `true`; `0` is `false`. Boolean expressions do not strictly require `bool` variables — integers work too.

---

### 1.5.1 `if` / `else if` / `else`

```c
if (boolean-expr1)
{
    // first branch
}
else if (boolean-expr2)
{
    // second branch
}
else if (boolean-expr3)
{
    // third branch
}
else
{
    // fourth branch — catch-all
}
```

In an `if / else if / else` chain, **all branches are mutually exclusive** — exactly one will execute. Once a true branch is found, the rest are skipped entirely.

The bare `else` handles all remaining cases without an extra check. Replacing it with `else if (x == y)` is wasteful — the condition is unnecessary.

---

### 1.5.2 Non-Mutually-Exclusive Chains

It is also possible to write a chain of independent `if` blocks that are **not** mutually exclusive:

```c
if (boolean-expr1)
{
    // first branch
}
if (boolean-expr2)
{
    // second branch
}
if (boolean-expr3)
{
    // third branch
}
else
{
    // fourth branch
}
```

**Key rule: `else` binds to the nearest `if` only.** In this example, only the third and fourth branches are mutually exclusive. The first and second `if` blocks run independently — multiple branches can fire in the same execution.

---

### 1.5.3 Comparison Operators

| Operator | Meaning |
|----------|---------|
| `<` | less than |
| `>` | greater than |
| `<=` | less than or equal to |
| `>=` | greater than or equal to |
| `==` | equal to (**two** equals signs) |
| `!=` | not equal to |

The most common C bug: using `=` (assignment) instead of `==` (comparison) inside a condition.

---

### 1.5.4 Logical Operators

| Operator | Name | Symbol |
|----------|------|--------|
| AND | true only if **both** operands are true | `&&` |
| OR | true if **at least one** operand is true | `\|\|` |
| NOT | inverts true/false | `!` |

**AND truth table:**

| x | y | x && y |
|---|---|--------|
| true | true | **true** |
| true | false | false |
| false | true | false |
| false | false | false |

**OR truth table:**

| x | y | x \|\| y |
|---|---|---------|
| true | true | **true** |
| true | false | **true** |
| false | true | **true** |
| false | false | false |

**NOT truth table:**

| x | !x |
|---|----|
| true | false |
| false | **true** |

`||` is typed as two vertical bar characters (usually on the same key as `\`, above Enter). Using `&&` where `||` is needed is a logic bug — a variable cannot simultaneously satisfy two mutually exclusive conditions.

---

### Arithmetic Operators

| Operator | Meaning | Notes |
|----------|---------|-------|
| `+` | addition | |
| `-` | subtraction | |
| `*` | multiplication | |
| `/` | division | truncates when both operands are `int` |
| `%` | modulo — remainder | `13 % 4` → `1` |

### `switch` Statement

`switch` selects among discrete, enumerated values without chaining boolean expressions. Preferred when comparing a single variable against many specific values.

```c
int x = get_int("Enter 1, 2, or 3: ");

switch (x)
{
    case 1:
        printf("One!\n");
        break;
    case 2:
        printf("Two!\n");
        break;
    case 3:
        printf("Three!\n");
        break;
    default:
        printf("Sorry!\n");
}
```

- `break` exits the switch — **without it, execution falls through to the next case**.
- `default` handles any value not matched by a `case`.

**Deliberate fall-through** — no `break`, intentional countdown:
```c
switch (n)
{
    case 5: printf("Five, ");
    case 4: printf("Four, ");
    case 3: printf("Three, ");
    case 2: printf("Two, ");
    case 1: printf("One, ");
    default: printf("Blastoff!\n");
}
```
If `n` is 3: prints `Three, Two, One, Blastoff!`

---

### 1.5.5 Ternary Operator `?:`

Shorthand for a simple `if/else` that assigns one of two values:

```c
// Standard if/else
int x;
if (expr) { x = 5; }
else      { x = 6; }

// Equivalent ternary
int x = (expr) ? 5 : 6;
```

Mostly a compact trick for trivially short branches. Not appropriate when branches contain multiple statements.

---

### 1.5.6 Summary: When to Use Each

| Construct | Use when |
|-----------|---------|
| `if / else if / else` | making decisions based on Boolean expressions |
| `switch` | comparing one variable against discrete known values |
| `?:` | replacing a very simple `if/else` with a single-line assignment |


## 1.6 Loops

Loops allow programs to execute lines of code repeatedly without copy-pasting them. Three types exist in C, all interchangeable in theory but each with a natural use case.

---

### 1.6.1 When to Use Each Loop

| Loop | Use when… | Runs at least once? |
|------|-----------|-------------------|
| `while` | unknown number of iterations; may not run at all | No |
| `do while` | unknown but **at least once** | **Yes** |
| `for` | discrete, known (or computable) number of times | Depends on condition |

---

### 1.6.2 `while` Loop

The condition is checked *before* each iteration. If false on the first check, the body never executes.

```c
while (boolean-expr)
{
    // executes repeatedly while boolean-expr is true
}
```

```c
int i = 0;
while (i < 3)
{
    printf("meow\n");
    i++;
}
```

**Scratch note**: C's `while` continues as long as the condition is **true**. Scratch's `repeat until` block continues until the condition is **true** (i.e., it repeats while the condition is *false*). The logic is reversed — be careful when translating from Scratch to C.

Forgetting `i++` inside produces an infinite loop — the condition never becomes false.

---

### 1.6.3 `for` Loop

Condenses initialization, condition, and update into one line. The general form:

```c
for (start; expr; increment)
{
    // body
}
```

Step-by-step execution:
1. `start` executes **once** at the beginning
2. `expr` is checked — if **false**, loop ends immediately
3. body executes
4. `increment` executes
5. back to step 2

```c
for (int i = 0; i < 3; i++)
{
    printf("meow\n");
}
```

A variable declared inside the `for` header (`int i`) only exists within that loop's scope. Two separate `for` loops can both use `int i` without conflict.

---

### 1.6.4 `do while` Loop

`do` is a keyword meaning **"execute the following block first, no questions asked."** The `while` condition comes *after* the closing `}` — at the bottom — so it is only checked once the body has already run.

```c
do
{
    // always runs at least once
}
while (boolean-expr);   // condition checked HERE, after the body
```

Compare to a regular `while`, where the condition is at the top:
- `while` — ask first, maybe never run the body
- `do while` — run the body first, then ask whether to repeat

```c
int n;
do
{
    n = get_int("Enter a positive number: ");
}
while (n < 1);
```

Step-by-step:
1. `do` — execute the body (prompt the user)
2. check `n < 1` — if true, go back to step 1; if false, exit the loop

This is the natural pattern for prompting user input: you obviously need to ask at least once before you can check whether the answer was valid.

---

### Infinite Loops, `break`, and `continue`

`while (true)` creates a deliberate infinite loop:

```c
while (true)
{
    n = get_int("Positive integer: ");
    if (n >= 1)
        break;       // exit the loop immediately
}
```

- `break` — exit the loop entirely; jump past the closing `}`
- `continue` — skip the rest of the current iteration; jump back to the condition check

Ctrl-C in the terminal kills a running program — the escape hatch for accidental infinite loops.

---

### 1.6.5 Scope

A variable only exists within the **curly braces** in which it was declared.

```c
int n;                          // declared outside — visible everywhere below
while (true)
{
    n = get_int("n? ");         // assigns to the outer n — OK
    if (n >= 0) break;
}
printf("%i\n", n);              // still accessible here
```

If `n` were declared *inside* the loop, the `printf` would produce an `undeclared identifier` error. Always declare the variable in the outermost scope where it is needed.

---

### 1.6.6 Nested Loops — Mario Grid

An outer loop over rows and an inner loop over columns produces a 2D grid:

```c
#define SIZE 3

for (int row = 0; row < SIZE; row++)
{
    for (int col = 0; col < SIZE; col++)
    {
        printf("#");
    }
    printf("\n");   // newline after each complete row — must be outside the inner loop
}
```

Placing `\n` *inside* the inner loop prints a column instead of a row. `i` and `j` are conventional counter names for nested loops; `k` for a third level. Beyond that, descriptive names are required.


## 1.7 Functions

A **function** is a named block of code that performs a task. Functions avoid repetition, make code readable, and allow the same logic to be reused without re-implementing it.

---

### 1.7.1 Defining a Function

```c
void meow(void)
{
    printf("meow\n");
}
```

- First `void`: the **return type** — this function returns nothing (has only a side effect).
- `meow`: the **name**.
- Second `void` in parentheses: the **parameters** — this function takes no input.

Calling it:
```c
meow();    // prints "meow"
```

---

### 1.7.2 Functions with Parameters

A parameter gives the function input it can use. The type and name of each parameter are declared inside the parentheses:

```c
void meow(int times)
{
    for (int i = 0; i < times; i++)
    {
        printf("meow\n");
    }
}
```

Calling it:
```c
meow(3);    // prints meow 3 times
meow(n);    // prints meow n times, where n is some variable
```

---

### 1.7.3 Functions with Return Values

When a function produces a value that should be used elsewhere, it returns that value. The return type replaces `void`:

```c
int get_positive_int(void)
{
    int n;
    do
    {
        n = get_int("Enter a positive integer: ");
    }
    while (n < 1);
    return n;    // hand the value back to the caller
}
```

Using it:
```c
int n = get_positive_int();   // n receives whatever get_positive_int returned
```

---

### 1.7.4 Prototypes

C compilers read files top to bottom. If `main` calls `meow` but `meow` is defined below `main`, the compiler produces an `undeclared function` error.

A **prototype** is a one-line declaration placed above `main` that promises the function will be defined later:

```c
void meow(int times);   // prototype — semicolon, no body

int main(void)
{
    meow(3);
}

void meow(int times)    // full definition — can be placed below main
{
    for (int i = 0; i < times; i++)
        printf("meow\n");
}
```

This is the only sanctioned use of copy-paste in C: copy the first line of the function definition, add a semicolon, and place it above `main`.

---

### 1.7.5 Scope in Functions

A variable declared in `main` is not visible inside other functions — each function has its own scope. The way to share a value across function calls is to **pass it as an argument**:

```c
int n = get_int("How many times? ");
meow(n);    // n is passed as input; meow sees it as 'times'
```

Attempting to use `n` directly inside `meow` without passing it would produce an `undeclared identifier` error.

---

### 1.7.6 Comments

A `//` comment is a note to the human, ignored entirely by the compiler:
```c
// get a non-negative integer from the user
int n = get_positive_int();

meow(n);    // meow that many times
```

Comments should explain *why*, not *what* (the code already says what it does).


## 1.8 Code Quality

Three axes are used to evaluate code, both in CS50 and in the real world:

| Axis | Question | Evaluated by |
|------|----------|--------------|
| **Correctness** | Does it do what it's supposed to? | `check50` (automated tests) |
| **Design** | Is it implemented efficiently and cleanly? | `design50` (AI feedback), TAs |
| **Style** | Is it consistently formatted and readable? | `style50` (automated formatting check) |

**Correctness** is necessary but not sufficient. Correct code can still be poorly designed or unreadable.

**Design** covers: avoiding redundant checks, not repeating code, choosing the right loop, naming variables meaningfully, not creating unnecessary intermediate variables.

**Style** covers: consistent indentation (4 spaces in CS50), brace placement, variable naming conventions. It matters to humans, not the compiler.

---

### 1.8.1 Design: `if/else if/else` vs. Three `if` Blocks

**Worse (checks all three conditions every time):**
```c
if (x < y)  { printf("x is less than y\n"); }
if (x > y)  { printf("x is greater than y\n"); }
if (x == y) { printf("x is equal to y\n"); }
```

**Better (short-circuits on first match):**
```c
if (x < y)
    printf("x is less than y\n");
else if (x > y)
    printf("x is greater than y\n");
else
    printf("x is equal to y\n");
```

---

### 1.8.2 Magic Numbers

A **magic number** is a numeric literal embedded directly in code with no explanation of what it means:

```c
// What does 52 mean? What does 23 mean?
for (int i = 0; i < 52; i++)
    deal_card(deck);
```

The height of Mario's pyramid is capped at 23 — because the standard terminal window is 24 lines tall. But that meaning is invisible from the code itself.

**Step 1 — named variable (better, but not safe enough):**
```c
int deck_size = 52;
for (int i = 0; i < deck_size; i++)
    deal_card(deck);
```

This gives the number a name, but `deck_size` is a variable — another function could inadvertently change it, causing unpredictable behavior.

**Step 2 — `#define` (best):**
```c
#define DECKSIZE 52

for (int i = 0; i < DECKSIZE; i++)
    deal_card(deck);
```

---

### 1.8.3 `#define` — Preprocessor Constants

`#define` is a **preprocessor directive** — also called a **macro**. Before compilation begins, the preprocessor performs a literal find-and-replace across the entire file:

```
#define NAME REPLACEMENT
```

> If `#include` is similar to copy/paste, then `#define` is analogous to find/replace.

Every occurrence of `NAME` in the code is replaced with `REPLACEMENT` before the compiler ever sees it.

```c
#define PI       3.14159265
#define COURSE   "CS50"
#define DECKSIZE 52
```

**Rules:**
- No semicolon at the end of a `#define` line.
- By convention, `#define` names are written in **ALL_CAPS** to distinguish them visually from variables.
- The value is a literal substitution — it cannot be accidentally changed at runtime.

**`#define` advantage over a named variable:** changing one `#define` line propagates everywhere automatically. Switching to a German deck (`DECKSIZE 32`) or an absurdly large one (`DECKSIZE 50000`) requires editing exactly one line:

```c
#define DECKSIZE 32   // change here only — propagates throughout all files
```

**`#define` vs `const`:**

| | `const int N = 52;` | `#define N 52` |
|-|---------------------|----------------|
| Type-checked by compiler | Yes | No (raw text substitution) |
| Can be accidentally changed | Yes (another function could do `N++`) | No (52 is a literal) |
| Has an address in memory | Yes | No |
| Convention | mixed case OK | ALL_CAPS |

`#define` is preferred for true symbolic constants. `const` is preferred when the type matters or the value is computed at runtime.

---

### 1.8.4 Design: Constants in Mario

Hard-coded `3` in two places — fragile:
```c
for (int row = 0; row < 3; row++)
{
    for (int col = 0; col < 3; col++)
        printf("#");
    printf("\n");
}
```

With `#define` — one line controls everything:
```c
#define SIZE 3

for (int row = 0; row < SIZE; row++)
{
    for (int col = 0; col < SIZE; col++)
        printf("#");
    printf("\n");
}
```

The ALL_CAPS name signals at a glance that `SIZE` is a fixed constant, not a variable that will change.


## 1.9 Numeric Limitations

Computers represent numbers in a finite number of bits. Two classes of problems arise from this.

---

### 1.9.1 Integer Overflow

A 32-bit integer can hold at most ~2 billion (2³¹ − 1 for signed integers). If a value exceeds this, the bits "wrap around" — the counter overflows back toward zero or into negative numbers.

```c
int dollars = 1;
while (true)
{
    printf("Here is $%i. Double it? ", dollars);
    char c = get_char("");
    if (c == 'y') dollars *= 2;
    else break;
}
```

After enough doublings (roughly 30), `dollars` goes negative, then to zero — the integer overflowed.

Using `long` (64 bits) delays the problem but does not eliminate it — the same overflow occurs after more doublings.

**Real-world consequences:**
- **Boeing 787 (2015)**: after 248 days of continuous power, the generator control units could lose all AC electrical power due to an internal software counter overflowing a 32-bit integer. The short-term fix was to reboot the plane before 248 days elapsed.
- **Pac-Man**: reaching level 256 garbled half the screen because the level counter, stored in a single byte (max 255), overflowed to 0, skipping a bounds check and writing random data to the display buffer.

---

### 1.9.2 Floating Point Imprecision

A `float` uses 32 bits to represent a real number. Since there are infinitely many real numbers but only ~4 billion possible 32-bit patterns, most values are approximations.

**Integer division truncates:**
```c
int x = 1, y = 3;
printf("%i\n", x / y);   // prints 0, not 0.333...
```
When dividing two integers in C, the result is an integer — everything after the decimal point is discarded. To get a decimal result, at least one operand must be a `float` or `double`.

**Floating point imprecision:**
```c
float x = 1.0, y = 3.0;
printf("%.50f\n", x / y);
// 0.33333334326744079589843750000000000000000000000000
```

1/3 cannot be represented exactly in binary floating point. The result is an approximation — correct for a few significant digits, then meaningless noise. Using `double` (64 bits) provides more precision but the same fundamental problem remains.

---

### 1.9.3 Y2K and the Year 2038 Problem

Both are practical consequences of representing numbers with too few bits.

**Y2K (Year 2000)**: for decades, years were stored as two digits (`99` for 1999) to save memory. When the year rolled to 2000, systems stored `00`, which many programs interpreted as 1900. A global scramble was required to patch software before the rollover.

**Year 2038**: many systems, following a convention established in the 1970s, track time as the number of seconds elapsed since January 1, 1970, stored in a 32-bit signed integer. That counter overflows on **January 19, 2038** — at which point affected systems may interpret the time as December 13, 1901.

The fix is to migrate to 64-bit time representations, which delays the overflow by hundreds of billions of years. Many modern systems have already made this change; older embedded systems remain at risk.


## 1.10 Problem Set

### 1.10.1 Mario Less

#### 1.10.1.1 The Problem

Implement a right-aligned pyramid using `#` characters. The height is provided by the user. Re-prompt if the input is not a positive integer.

Example output for height 3:
```
  #
 ##
###
```

---

#### 1.10.1.2 How the pyramid works

For a pyramid of height `n`, each row `i` (starting at 0) prints:
- `n - 1 - i` spaces (decreases each row)
- `i + 1` hashes (increases each row)

As `i` goes up by 1, one less space and one more hash — that's what creates the diagonal staircase.

---

#### 1.10.1.3 Issues faced

**Bug 1 — outer loop only ran once:**
```c
for (int i = 0; i < 1; i++)  // wrong: always stops after one row
for (int i = 0; i < n; i++)  // correct: runs n rows
```

**Bug 2 — wrong number of hashes:**
```c
for (int j = 0; j < n + 1; j++)  // wrong: prints n+1 hashes on every row
for (int j = 0; j < i + 1; j++)  // correct: grows with each row
```

**Bug 3 — no re-prompting for invalid input:**
```c
int n = get_int("Height: ");  // wrong: accepts 0 or negative with no complaint
```

The fix is a `do while` loop. It asks at least once, then checks — if the answer is invalid, it asks again:
```c
int n;
do
{
    n = get_int("Height: ");
}
while (n < 1);
```

`n` must be declared *outside* the loop so it's still accessible after the loop ends. A regular `while` won't work here because you need to ask at least once before you can check anything.

---

#### Final correct code

```c
#include <stdio.h>
#include <cs50.h>

void bricks(int n)
{
    for (int i = 0; i < n; i++)
    {
        // print steps
        for (int j = 0; j < n - 1 - i; j++)
            printf(" ");
        // lay bricks
        for (int j = 0; j < i + 1; j++)
            printf("#");
        printf("\n");
    }
}

int main(void)
{
    int n;
    do
    {
        n = get_int("Height: ");
    }
    while (n < 1);
    bricks(n);
}
```

---

#### Chronological execution: what happens when `./mario` runs

**Step 1 — program enters `main`**
The OS hands control to `main`. The first thing it sees is `int n;` — this reserves a slot in memory for an integer called `n`. No value is assigned yet; whatever is in that memory slot is garbage.

**Step 2 — `do` keyword**
`do` is an instruction to the computer: *execute the block below immediately, without checking any condition first.* There is no gate — the body runs unconditionally the first time.

**Step 3 — `get_int("Height: ")` runs**
The program prints `Height: ` and waits. The user types something. `get_int` has its own internal rules:
- If the input is not an integer (e.g. `foo`, `3.5`), it rejects it and re-prompts on its own — the programmer does not need to handle this.
- Once a valid integer is entered, `get_int` returns it and assigns it to `n`.

**Step 4 — `while (n < 1)` is checked**
This is the rule the programmer added on top of `get_int`'s built-in checks. It asks: *is the number the user gave less than 1?*
- If **yes** (e.g. `0`, `-3`): the condition is true → loop back to step 2, run `do` block again, re-prompt.
- If **no** (e.g. `3`): the condition is false → exit the loop and continue.

This back-and-forth between steps 2–4 repeats until the user gives a valid positive integer.

**Step 5 — `bricks(n)` is called**
`main` passes the validated value of `n` into `bricks`. Inside `bricks`, the parameter is also called `n` and holds that same value (e.g. `3`).

**Step 6 — outer `for` loop starts (`i = 0`)**
The `for` loop sets up three rules at once:
- *start*: `i = 0` — runs once, sets the counter
- *condition*: `i < n` — checked before every iteration; if false, loop ends
- *increment*: `i++` — runs after every iteration, advances the counter

First check: `0 < 3` → true → enter the body.

**Step 7 — inner loop 1: spaces**
`for (int j = 0; j < n - 1 - i; j++)` — same three-rule structure.
With `n = 3` and `i = 0`: condition is `j < 2`, so it runs twice → prints 2 spaces.

**Step 8 — inner loop 2: hashes**
`for (int j = 0; j < i + 1; j++)` — with `i = 0`: condition is `j < 1`, runs once → prints 1 hash.

**Step 9 — `printf("\n")`**
Moves the cursor to the next line. Row 0 is complete: `  #`

**Step 10 — `i++`, condition re-checked**
`i` becomes `1`. Check `1 < 3` → true → repeat steps 7–9 with `i = 1`:
- spaces: `j < 1` → 1 space
- hashes: `j < 2` → 2 hashes
- result: ` ##`

**Step 11 — `i = 2`, repeat**
- spaces: `j < 0` → 0 spaces (loop body never runs)
- hashes: `j < 3` → 3 hashes
- result: `###`

**Step 12 — `i++` → `i = 3`, condition fails**
`3 < 3` is false → outer loop exits. `bricks` returns. `main` returns. Program ends.


### 1.10.2 Mario More

#### 1.10.2.1 The Problem

Same as Mario Less, but now print **two pyramids side by side** with a 2-space gap between them. Height must be between 1 and 8 inclusive.

Example output for height 4:
```
   #  #
  ##  ##
 ###  ###
####  ####
```

---

#### 1.10.2.2 What changed from Mario Less

**1. The `while` condition is stricter**

Mario Less only rejected below 1:
```c
while (n < 1);
```

Mario More also rejects above 8:
```c
while (n < 1 || n > 8);
```

`||` means OR — the loop keeps re-prompting if *either* condition is true: too small or too big.

**2. `bricks` prints a gap and a second pyramid per row**

Mario Less printed one row like this:
```
[spaces][left hashes]\n
```

Mario More prints:
```
[spaces][left hashes][2-space gap][right hashes]\n
```

The left side is identical to Mario Less. The gap is always exactly 2 spaces regardless of height. The right side is the same number of hashes as the left — no leading spaces needed because the gap already positions it.

---

#### Final correct code

```c
#include <cs50.h>
#include <stdio.h>

void bricks(int n)
{
    for (int i = 0; i < n; i++)
    {
        // print steps
        for (int j = 0; j < n - 1 - i; j++)
            printf(" ");
        // lay left bricks
        for (int j = 0; j < i + 1; j++)
            printf("#");
        // print gap
        printf("  ");
        // lay right bricks
        for (int j = 0; j < i + 1; j++)
            printf("#");
        printf("\n");
    }
}

int main(void)
{
    int n;
    do
    {
        n = get_int("Height: ");
    }
    while (n < 1 || n > 8);
    bricks(n);
}
```

**Note on reusing `j`**: all three inner loops use `j` as their counter. This is fine because each loop is sequential — `j` is created at the start of each `for` and ceases to exist when that loop ends, before the next one begins. There is no overlap.

---

#### Chronological execution: what happens when `./mario` runs

**Step 1 — program enters `main`**
`int n;` reserves a slot in memory for `n`. No value yet.

**Step 2 — `do` keyword**
Same as Mario Less: execute the block below immediately without checking any condition first.

**Step 3 — `get_int("Height: ")` runs**
Prints `Height: ` and waits. `get_int`'s own internal rules reject anything that isn't an integer (letters, decimals, empty input) and re-prompt automatically.

**Step 4 — `while (n < 1 || n > 8)` is checked**
Two rules the programmer added on top:
- `n < 1` — rejects 0 and negatives
- `n > 8` — rejects 9 and above
- `||` — if *either* is true, loop back to step 2 and ask again

Only when both are false (i.e. `n` is between 1 and 8) does the loop exit.

**Step 5 — `bricks(n)` is called**
The validated value of `n` is passed in (e.g. `4`).

**Step 6 — outer `for` loop starts (`i = 0`)**
Three rules set up at once:
- `i = 0` — runs once to initialise
- `i < n` — checked before every row; false means stop
- `i++` — runs after every row to advance

First check: `0 < 4` → true → enter the body.

**Step 7 — inner loop 1: leading spaces**
`for (int j = 0; j < n - 1 - i; j++)` — with `n = 4`, `i = 0`: condition is `j < 3`, runs 3 times → prints 3 spaces. Each row gets one fewer space as `i` grows.

**Step 8 — inner loop 2: left hashes**
`for (int j = 0; j < i + 1; j++)` — with `i = 0`: condition is `j < 1`, runs once → prints 1 `#`. Each row gets one more hash as `i` grows.

**Step 9 — gap**
`printf("  ")` — always prints exactly 2 spaces. No loop, no condition — just a fixed separator. This never changes regardless of height.

**Step 10 — inner loop 3: right hashes**
`for (int j = 0; j < i + 1; j++)` — identical rule to inner loop 2. With `i = 0`: prints 1 `#`. The right pyramid is always a mirror of the left.

**Step 11 — `printf("\n")`**
Ends the row. Row 0 complete: `   #  #`

**Step 12 — `i++`, condition re-checked**
`i` becomes `1`. Check `1 < 4` → true → repeat steps 7–11:
- spaces: `j < 2` → 2 spaces
- left hashes: `j < 2` → 2 hashes
- gap: `  `
- right hashes: `j < 2` → 2 hashes
- result: `  ##  ##`

**Step 13 — continues through `i = 2` and `i = 3`**

`i = 2`: ` ###  ###`
`i = 3`: `####  ####`

**Step 14 — `i++` → `i = 4`, condition fails**
`4 < 4` is false → outer loop exits. `bricks` returns. `main` returns. Program ends.


## 1.11 Errata / Local Setup Notes

Issues encountered running CS50 material locally (outside the Harvard Codespace) and their fixes.

---

### 1.11.1 CS50 Library (`cs50.h`) not found on Ubuntu 26.04

**Error:**
```
hello.c:1:10: fatal error: cs50.h: No such file or directory
```

**Cause:** The CS50 packagecloud repo doesn't have builds for Ubuntu 26.04 yet, so `sudo apt install libcs50` fails.

**Fix:** Install from source, but **not** from a Windows filesystem path. WSL's `/mnt/c/...` doesn't support Linux file permissions, causing `git clone` to fail with `chmod` errors. Clone into the Linux home directory instead:

```bash
cd ~
git clone https://github.com/cs50/libcs50
cd libcs50
sudo make install
```

---

### 1.11.2 `make` doesn't link the CS50 library

**Error:**
```
undefined reference to `get_string'
```

**Cause:** `make` without a Makefile uses a default rule that doesn't know to link `-lcs50`, so any function from `cs50.h` (e.g. `get_string`, `get_int`) causes a linker error.

**Fix:** Compile manually with `gcc` and pass the library flag explicitly:

```bash
gcc hello.c -o hello -lcs50
./hello
```
